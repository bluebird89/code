````python
# %load https://www.dataquest.io/blog/jupyter-notebook-tutorial/
<!doctype html>
<html lang="en-US">
<head>
	<link rel="profile" href="https://gmpg.org/xfn/11">
	<meta charset="UTF-8">
	<meta name="viewport" content="width=device-width, initial-scale=1">
		
	<!-- This site is optimized with the Yoast SEO Premium plugin v14.0.4 - https://yoast.com/wordpress/plugins/seo/ -->
	<title>Jupyter Notebook for Beginners Tutorial — Dataquest</title>
	<meta name="description" content="Use this tutorial to learn how to create your first Jupyter Notebook, important terminology, and how easily notebooks can be shared and published online." />
	<meta name="robots" content="index, follow" />
	<meta name="googlebot" content="index, follow, max-snippet:-1, max-image-preview:large, max-video-preview:-1" />
	<meta name="bingbot" content="index, follow, max-snippet:-1, max-image-preview:large, max-video-preview:-1" />
	<link rel="canonical" href="https://www.dataquest.io/blog/jupyter-notebook-tutorial/" />
	<meta property="og:locale" content="en_US" />
	<meta property="og:type" content="article" />
	<meta property="og:title" content="Jupyter Notebook for Beginners Tutorial — Dataquest" />
	<meta property="og:description" content="Use this tutorial to learn how to create your first Jupyter Notebook, important terminology, and how easily notebooks can be shared and published online." />
	<meta property="og:url" content="https://www.dataquest.io/blog/jupyter-notebook-tutorial/" />
	<meta property="og:site_name" content="Dataquest" />
	<meta property="article:publisher" content="https://www.facebook.com/dataquestio" />
	<meta property="article:published_time" content="2019-08-22T07:01:00+00:00" />
	<meta property="article:modified_time" content="2020-02-04T15:22:27+00:00" />
	<meta property="og:image" content="https://www.dataquest.io/wp-content/uploads/2019/01/1-LPnY8nOLg4S6_TG0DEXwsg-1.png" />
	<meta property="og:image:width" content="1600" />
	<meta property="og:image:height" content="795" />
	<meta name="twitter:card" content="summary_large_image" />
	<meta name="twitter:creator" content="@dataquestio" />
	<meta name="twitter:site" content="@dataquestio" />
	<script type="application/ld+json" class="yoast-schema-graph">{"@context":"https://schema.org","@graph":[{"@type":"Organization","@id":"https://www.dataquest.io/#organization","name":"Dataquest","url":"https://www.dataquest.io/","sameAs":["https://www.facebook.com/dataquestio","https://www.linkedin.com/company/dataquest-io","https://twitter.com/dataquestio"],"logo":{"@type":"ImageObject","@id":"https://www.dataquest.io/#logo","inLanguage":"en-US","url":"https://www.dataquest.io/wp-content/uploads/2019/01/Logo-On-Light-BLUE.png","width":2001,"height":427,"caption":"Dataquest"},"image":{"@id":"https://www.dataquest.io/#logo"}},{"@type":"WebSite","@id":"https://www.dataquest.io/#website","url":"https://www.dataquest.io/","name":"Dataquest","description":"Aquire the skills you need to start and advance your data science career.","publisher":{"@id":"https://www.dataquest.io/#organization"},"potentialAction":[{"@type":"SearchAction","target":"https://www.dataquest.io/?s={search_term_string}","query-input":"required name=search_term_string"}],"inLanguage":"en-US"},{"@type":"ImageObject","@id":"https://www.dataquest.io/blog/jupyter-notebook-tutorial/#primaryimage","inLanguage":"en-US","url":"https://www.dataquest.io/wp-content/uploads/2019/01/1-LPnY8nOLg4S6_TG0DEXwsg-1.png","width":1600,"height":795,"caption":"jupyter-notebook best free data science tools"},{"@type":"WebPage","@id":"https://www.dataquest.io/blog/jupyter-notebook-tutorial/#webpage","url":"https://www.dataquest.io/blog/jupyter-notebook-tutorial/","name":"Jupyter Notebook for Beginners Tutorial \u2014 Dataquest","isPartOf":{"@id":"https://www.dataquest.io/#website"},"primaryImageOfPage":{"@id":"https://www.dataquest.io/blog/jupyter-notebook-tutorial/#primaryimage"},"datePublished":"2019-08-22T07:01:00+00:00","dateModified":"2020-02-04T15:22:27+00:00","description":"Use this tutorial to learn how to create your first Jupyter Notebook, important terminology, and how easily notebooks can be shared and published online.","inLanguage":"en-US","potentialAction":[{"@type":"ReadAction","target":["https://www.dataquest.io/blog/jupyter-notebook-tutorial/"]}]},{"@type":"Article","@id":"https://www.dataquest.io/blog/jupyter-notebook-tutorial/#article","isPartOf":{"@id":"https://www.dataquest.io/blog/jupyter-notebook-tutorial/#webpage"},"author":{"@id":"https://www.dataquest.io/#/schema/person/73a2d9199da8ee0fc17ecd6c3923c334"},"headline":"Jupyter Notebook for Beginners: A Tutorial","datePublished":"2019-08-22T07:01:00+00:00","dateModified":"2020-02-04T15:22:27+00:00","commentCount":0,"mainEntityOfPage":{"@id":"https://www.dataquest.io/blog/jupyter-notebook-tutorial/#webpage"},"publisher":{"@id":"https://www.dataquest.io/#organization"},"image":{"@id":"https://www.dataquest.io/blog/jupyter-notebook-tutorial/#primaryimage"},"keywords":"beginner,Jupyter,jupyter notebooks,Learn Python,Pandas,python,tutorial,Tutorials","articleSection":"Data Science Tutorials","inLanguage":"en-US"},{"@type":["Person"],"@id":"https://www.dataquest.io/#/schema/person/73a2d9199da8ee0fc17ecd6c3923c334","name":"Benjamin Pryke","image":{"@type":"ImageObject","@id":"https://www.dataquest.io/#personlogo","inLanguage":"en-US","url":"https://www.dataquest.io/wp-content/uploads/2015/12/benjamin_pryke_profile.jpg","caption":"Benjamin Pryke"},"description":"Python and web developer with a background in computer science and machine learning. Co-founder of FinTech firm Machina Capital. Part-time gymnast and digital bohemian."}]}</script>
	<!-- / Yoast SEO Premium plugin. -->


<link rel='dns-prefetch' href='//s.w.org' />
<link rel="alternate" type="application/rss+xml" title="Dataquest &raquo; Feed" href="https://www.dataquest.io/feed/" />
<link rel="alternate" type="application/rss+xml" title="Dataquest &raquo; Comments Feed" href="https://www.dataquest.io/comments/feed/" />
		<script type="text/javascript">
			window._wpemojiSettings = {"baseUrl":"https:\/\/s.w.org\/images\/core\/emoji\/12.0.0-1\/72x72\/","ext":".png","svgUrl":"https:\/\/s.w.org\/images\/core\/emoji\/12.0.0-1\/svg\/","svgExt":".svg","source":{"concatemoji":"https:\/\/www.dataquest.io\/wp-includes\/js\/wp-emoji-release.min.js?ver=5.4.1"}};
			/*! This file is auto-generated */
			!function(e,a,t){var r,n,o,i,p=a.createElement("canvas"),s=p.getContext&&p.getContext("2d");function c(e,t){var a=String.fromCharCode;s.clearRect(0,0,p.width,p.height),s.fillText(a.apply(this,e),0,0);var r=p.toDataURL();return s.clearRect(0,0,p.width,p.height),s.fillText(a.apply(this,t),0,0),r===p.toDataURL()}function l(e){if(!s||!s.fillText)return!1;switch(s.textBaseline="top",s.font="600 32px Arial",e){case"flag":return!c([127987,65039,8205,9895,65039],[127987,65039,8203,9895,65039])&&(!c([55356,56826,55356,56819],[55356,56826,8203,55356,56819])&&!c([55356,57332,56128,56423,56128,56418,56128,56421,56128,56430,56128,56423,56128,56447],[55356,57332,8203,56128,56423,8203,56128,56418,8203,56128,56421,8203,56128,56430,8203,56128,56423,8203,56128,56447]));case"emoji":return!c([55357,56424,55356,57342,8205,55358,56605,8205,55357,56424,55356,57340],[55357,56424,55356,57342,8203,55358,56605,8203,55357,56424,55356,57340])}return!1}function d(e){var t=a.createElement("script");t.src=e,t.defer=t.type="text/javascript",a.getElementsByTagName("head")[0].appendChild(t)}for(i=Array("flag","emoji"),t.supports={everything:!0,everythingExceptFlag:!0},o=0;o<i.length;o++)t.supports[i[o]]=l(i[o]),t.supports.everything=t.supports.everything&&t.supports[i[o]],"flag"!==i[o]&&(t.supports.everythingExceptFlag=t.supports.everythingExceptFlag&&t.supports[i[o]]);t.supports.everythingExceptFlag=t.supports.everythingExceptFlag&&!t.supports.flag,t.DOMReady=!1,t.readyCallback=function(){t.DOMReady=!0},t.supports.everything||(n=function(){t.readyCallback()},a.addEventListener?(a.addEventListener("DOMContentLoaded",n,!1),e.addEventListener("load",n,!1)):(e.attachEvent("onload",n),a.attachEvent("onreadystatechange",function(){"complete"===a.readyState&&t.readyCallback()})),(r=t.source||{}).concatemoji?d(r.concatemoji):r.wpemoji&&r.twemoji&&(d(r.twemoji),d(r.wpemoji)))}(window,document,window._wpemojiSettings);
		</script>
		<style type="text/css">
img.wp-smiley,
img.emoji {
	display: inline !important;
	border: none !important;
	box-shadow: none !important;
	height: 1em !important;
	width: 1em !important;
	margin: 0 .07em !important;
	vertical-align: -0.1em !important;
	background: none !important;
	padding: 0 !important;
}
</style>
	<link rel='stylesheet' id='wp-block-library-css'  href='https://www.dataquest.io/wp-includes/css/dist/block-library/style.min.css?ver=5.4.1' type='text/css' media='all' />
<link rel='stylesheet' id='searchandfilter-css'  href='https://www.dataquest.io/wp-content/plugins/search-filter/style.css?ver=1' type='text/css' media='all' />
<link rel='stylesheet' id='tve_style_family_tve_flt-css'  href='https://www.dataquest.io/wp-content/plugins/thrive-visual-editor/editor/css/thrive_flat.css?ver=2.5.3.3' type='text/css' media='all' />
<link rel='stylesheet' id='tve_leads_forms-css'  href='//www.dataquest.io/wp-content/plugins/thrive-leads/editor-layouts/css/frontend.css?ver=2.2.14.3' type='text/css' media='all' />
<link rel='stylesheet' id='l2h_style-css'  href='https://www.dataquest.io/wp-content/plugins/latex2html/inc/css/latex.min.css?ver=2.3.7' type='text/css' media='screen, print' />
<link rel='stylesheet' id='l2h_print_style-css'  href='https://www.dataquest.io/wp-content/plugins/latex2html/inc/css/print.min.css?ver=2.3.7' type='text/css' media='print' />
<link rel='stylesheet' id='jquery-data-tables-css'  href='https://www.dataquest.io/wp-content/plugins/posts-data-table/assets/js/datatables/datatables.min.css?ver=1.10.18' type='text/css' media='all' />
<link rel='stylesheet' id='posts-data-table-css'  href='https://www.dataquest.io/wp-content/plugins/posts-data-table/assets/css/posts-data-table.min.css?ver=1.3' type='text/css' media='all' />
<link rel='stylesheet' id='thrive-css'  href='https://www.dataquest.io/wp-content/themes/thrive-theme/style.css?ver=1.2.3' type='text/css' media='all' />
<link rel='stylesheet' id='thrive-template-css'  href='//www.dataquest.io/wp-content/uploads/thrive/theme-template-1588724118.css?ver=5.4.1' type='text/css' media='all' />
<link rel='stylesheet' id='prism-theme-css'  href='https://www.dataquest.io/wp-content/plugins/ank-prism-for-wp/out/prism-css.min.css?ver=1550607458' type='text/css' media='all' />
<script type='text/javascript'>
/* <![CDATA[ */
var THO_Head = {"variations":[],"post_id":"2024","element_tag":"thrive_headline","woo_tag":"tho_woo"};
/* ]]> */
</script>
<script type='text/javascript' src='https://www.dataquest.io/wp-content/plugins/thrive-headline-optimizer/frontend/js/header.min.js?ver=1.2.7'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-includes/js/jquery/jquery.js?ver=1.12.4-wp'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-includes/js/jquery/jquery-migrate.min.js?ver=1.4.1'></script>
<link rel='https://api.w.org/' href='https://www.dataquest.io/wp-json/' />
<link rel="EditURI" type="application/rsd+xml" title="RSD" href="https://www.dataquest.io/xmlrpc.php?rsd" />
<link rel="wlwmanifest" type="application/wlwmanifest+xml" href="https://www.dataquest.io/wp-includes/wlwmanifest.xml" /> 
<meta name="generator" content="WordPress 5.4.1" />
<link rel='shortlink' href='https://www.dataquest.io/?p=2024' />
<link rel="alternate" type="application/json+oembed" href="https://www.dataquest.io/wp-json/oembed/1.0/embed?url=https%3A%2F%2Fwww.dataquest.io%2Fblog%2Fjupyter-notebook-tutorial%2F" />
<link rel="alternate" type="text/xml+oembed" href="https://www.dataquest.io/wp-json/oembed/1.0/embed?url=https%3A%2F%2Fwww.dataquest.io%2Fblog%2Fjupyter-notebook-tutorial%2F&#038;format=xml" />
<style type="text/css" id="tve_global_variables">:root{--tcb-color-0:rgb(35, 42, 61);--tcb-color-1:rgb(132, 132, 157);--tcb-color-2:rgb(77, 127, 223);--tcb-color-3:rgb(97, 209, 153);--tcb-color-4:rgb(0, 0, 0);--tcb-color-5:rgb(0, 0, 0);--tcb-color-6:rgb(0, 0, 0);--tcb-color-7:rgb(212, 212, 212);--tcb-color-8:rgb(137, 138, 142);--tcb-color-9:rgb(242, 243, 247);--tcb-color-10:rgb(35, 42, 61);--tcb-color-11:rgb(57, 207, 144);--tcb-color-12:rgb(255, 255, 255);--tcb-color-13:rgb(38, 53, 85);--tcb-color-14:rgb(73, 118, 234);--tcb-color-15:rgb(57, 207, 144);--tcb-color-16:rgb(42, 84, 167);--tcb-gradient-0:linear-gradient(90deg, rgb(97, 209, 153) 0%, rgb(78, 126, 222) 98%);--tcb-gradient-1:radial-gradient(var(--tcb-color-9) 0%, rgb(78, 126, 222) 76%);--tcb-gradient-2:linear-gradient(84deg, rgb(176, 97, 250) 0%, rgb(132, 143, 245) 18%, rgb(97, 180, 241) 42%, rgb(57, 207, 144) 100%);--tcb-skin-color-0:rgb(73, 118, 234);--tcb-skin-color-1:rgba(75, 115, 216, 0.5);--tcb-skin-color-3:rgb(24, 12, 69);--tcb-skin-color-8:rgba(75, 115, 216, 0.18);--tcb-skin-color-9:rgb(251, 252, 254);--tcb-skin-color-10:rgb(180, 223, 249);--tcb-skin-color-13:rgba(36, 72, 163, 0.5);--tcb-skin-color-16:rgb(98, 99, 100);--tcb-skin-color-18:rgba(68, 115, 233, 0.8);--tcb-skin-color-19:rgba(251, 252, 254, 0.85);--tcb-skin-color-20:rgba(251, 252, 254, 0.9);--tcb-skin-color-2:rgb(248, 248, 248);--tcb-skin-color-4:rgb(12, 17, 21);--tcb-skin-color-5:rgb(106, 107, 108);--tcb-skin-color-14:rgb(230, 230, 230);}</style>	<style>.async-hide { opacity: 0 !important} </style>
	<script>(function(a,s,y,n,c,h,i,d,e){s.className+=' '+y;h.start=1*new Date;
	h.end=i=function(){s.className=s.className.replace(RegExp(' ?'+y),'')};
	(a[n]=a[n]||[]).hide=h;setTimeout(function(){i();h.end=null},c);h.timeout=c;
	})(window,document.documentElement,'async-hide','dataLayer',2000,
	{'GTM-PNQ6824':true});</script>
	<script>
		/************* Helper functions ******************/
		function isInArray(value, array) {
		  return array.indexOf(value) > -1;
		}
		function getHostFromUrl(url) {
			var a=document.createElement('a');
			a.href=url;
			return a.hostname;
		}
		function getUrlVar(key){
			var result = new RegExp(key + "=([^&]*)", "i").exec(window.location.search); 
			return result && unescape(result[1]) || ""; 
		}
		/************* End of Helper Functions ******************/
		
		
		/******************** Cookie helper functions ************************/
		function checkCookieExists(cookie_name) {
			return document.cookie.indexOf(cookie_name + '=') !== -1;
		}
		function setCookie(name,value,domain,days) {
			var expires = "";
			if (days) {
				var date = new Date();
				date.setTime(date.getTime() + (days*24*60*60*1000));
				expires = "; expires=" + date.toUTCString();
			}
			document.cookie = name + "=" + (value || "")  + expires +  ";domain=" + domain + "; path=/";
		}
		/******************** End Of Cookie helper functions ******************/
		
		function setRefererUrlCookie() {
			var cookie_domain = '.dataquest.io'
			var internal_hosts =  ['dq-staging.t79ae38x-liquidwebsites.com', 'dataquest.io', ]

			/* Set Landing page cookie if it doesnot exist */
			if (!checkCookieExists('dataquest-cac-landing-page')) {
				setCookie('dataquest-cac-landing-page', window.location.pathname, cookie_domain, 7)
			}
			
			var referrer = document.referrer
			if (!referrer) {
				return
			}
			host = getHostFromUrl(referrer)
			
			/* If referrer is not coming from dataquest, set it in cookie and override landing_page cookie */
			if (!isInArray(host, internal_hosts)) {
				setCookie('dataquest-cac-referer-url', referrer, cookie_domain, 365)
				setCookie('dataquest-cac-landing-page', window.location.pathname, cookie_domain, 365)
			}
		}

		function setUTMCodesCookies() {
			var utm_codes = ['utm_source','utm_campaign','utm_content','utm_term','utm_medium']
			var cookie_prefix = 'dataquest-cac-'
			var cookie_domain = '.dataquest.io'
			
			for (code in utm_codes) {
				var utm_code = utm_codes[code]
				var code_value = getUrlVar(utm_code)
				var cookie_name = cookie_prefix + utm_code.replace('_', '-')
				
				if (code_value) {
					setCookie(cookie_name, code_value, cookie_domain, 365)
				}
			}
		}
									
		
		function setLandingPageLastTouchCookie() {
			var cookie_domain = '.dataquest.io'
			setCookie('dataquest-cac-landing-page-last-touch', window.location.pathname, cookie_domain, 365)

		}
		
		setUTMCodesCookies()
		setRefererUrlCookie()
		setLandingPageLastTouchCookie()

	</script>
<style>
.latex_thm, .latex_lem, .latex_cor, .latex_defn, .latex_prop, .latex_rem{
  margin:0;padding:5px;
  background: lightcyan;
  border: solid 3px green;
  -moz-border-radius: 1.0em;
  -webkit-border-radius: 7px;
  box-shadow: 0 0 0 green;
}
.latex_em{
  font-style: italic;
}
.bibtex_title{
  font-weight:bold;
  color: #004b33;
}
a.bibtex_title{
  text-decoration: none;
}
.latex_proof::after{
  content: "\220E";
  color: gray;
  text-align: right;
  display: block;
  font-size: 1.2em;
}
</style>
<script>
  !function(){var analytics=window.analytics=window.analytics||[];if(!analytics.initialize)if(analytics.invoked)window.console&&console.error&&console.error("Segment snippet included twice.");else{analytics.invoked=!0;analytics.methods=["trackSubmit","trackClick","trackLink","trackForm","pageview","identify","reset","group","track","ready","alias","debug","page","once","off","on"];analytics.factory=function(t){return function(){var e=Array.prototype.slice.call(arguments);e.unshift(t);analytics.push(e);return analytics}};for(var t=0;t<analytics.methods.length;t++){var e=analytics.methods[t];analytics[e]=analytics.factory(e)}analytics.load=function(t,e){var n=document.createElement("script");n.type="text/javascript";n.async=!0;n.src="https://cdn.segment.com/analytics.js/v1/"+t+"/analytics.min.js";var a=document.getElementsByTagName("script")[0];a.parentNode.insertBefore(n,a);analytics._loadOptions=e};analytics.SNIPPET_VERSION="4.1.0";
  analytics.load("BJhKYEXejzjqXq7xQDFI6N0M5W7HBAH5");
  analytics.page();
  }}();
</script>

<link rel="stylesheet" href="https://use.typekit.net/nmj4kuc.css"><style type="text/css" class="tve_global_style">@import url("https://fonts.googleapis.com/css?family=Literata:400,400i,600,600i,700,700i&subset=latin");@import url("//fonts.googleapis.com/css?family=Open+Sans:400,600,300,700,800&subset=latin");@import url("//fonts.googleapis.com/css?family=Open+Sans:400,600&subset=latin");@import url("//fonts.googleapis.com/css?family=Open+Sans:400,600,800&subset=latin");@import url("https://fonts.googleapis.com/css?family=Muli:400,400i,800,800i,700,700i,500,500i&subset=latin");@media(min-width: 300px){.tcb-global-button-k8mi4xps .tcb-button-link, .tcb-global-button-k8mi4xps button { border-radius: 5px; overflow: hidden; letter-spacing: var(--tve-letter-spacing, 2px); background-color: rgb(57, 207, 144)  !important; padding: 18px !important; background-image: none !important; }.tcb-global-button-k8mi4xps { --tve-font-size:16px; --tve-font-weight:var(--g-regular-weight, normal); --tve-letter-spacing:2px; --tcb-local-color-493ef:rgb(44, 168, 116); --tve-font-family:"Open Sans"; --g-regular-weight:400; --g-bold-weight:600; --tve-color:rgb(255, 255, 255); }.tcb-global-button-k8mi4xps .tcb-button-link span, .tcb-global-button-k8mi4xps button { font-weight: var(--tve-font-weight, var(--g-regular-weight, normal)); }.tcb-global-button-k8mi4xps:hover .tcb-button-link, .tcb-global-button-k8mi4xps:hover button { background-image: linear-gradient(rgba(255, 255, 255, 0.08), rgba(255, 255, 255, 0.08)) !important; background-size: auto !important; background-position: 50% 50% !important; background-attachment: scroll !important; background-repeat: no-repeat !important; background-color: var(--tcb-local-color-493ef)  !important; }.tcb-global-button-k8mi4xps-prtext { font-weight: var(--g-bold-weight, bold)  !important; }:not(#tve) .tcb-global-button-k8mi4xps .tcb-button-link span, :not(#tve) .tcb-global-button-k8mi4xps button { font-weight: var(--tve-font-weight, var(--g-regular-weight, normal)); color: var(--tve-color, rgb(255, 255, 255)); --tcb-applied-color:var$(--tcb-color-12); }:not(#tve) .tcb-global-button-k8mi4xps .tcb-button-link, :not(#tve) .tcb-global-button-k8mi4xps button { font-family: var(--tve-font-family, "Open Sans"); }:not(#tve) .tcb-global-button-k8mi4xps:hover .tcb-button-link, :not(#tve) .tcb-global-button-k8mi4xps:hover button { background-image: none !important; }}@media(max-width: 1023px){.tcb-global-button-k8mi4xps .tcb-button-link, .tcb-global-button-k8mi4xps button { padding: 16px !important; }}@media(max-width: 767px){.tcb-global-button-k8mi4xps .tcb-button-link, .tcb-global-button-k8mi4xps button { font-size: var(--tve-font-size, 16px)  !important; }.tcb-global-button-k8mi4xps { --tve-font-size:16px; }}</style><link type="text/css" rel="stylesheet" class="thrive-external-font" href="https://fonts.googleapis.com/css?family=Open+Sans:400,600,300,700,800&subset=latin&display=swap"><link type="text/css" rel="stylesheet" class="thrive-external-font" href="https://fonts.googleapis.com/css?family=Literata:400,400i,600,600i,700,700i&subset=latin&display=swap"><link type="text/css" rel="stylesheet" class="thrive-external-font" href="https://fonts.googleapis.com/css?family=Muli:400,400i,800,800i,700,700i,500,500i,600,300,200,900,300i,600i,900i,200i&subset=latin&display=swap"><style type="text/css" id="thrive-default-styles">@media (min-width: 300px) { .tcb-style-wrap h1{ color:var(--tcb-skin-color-4);--tcb-applied-color:var$(--tcb-skin-color-4);font-family:Literata;--g-regular-weight:400;font-weight:var(--g-bold-weight, bold);font-size:40px;background-color:rgba(0, 0, 0, 0);font-style:normal;margin:0px;padding-top:0px;padding-bottom:10px;text-decoration:rgba(10, 10, 10, 0.85);text-transform:none;border-left:0px none rgba(10, 10, 10, 0.85);--g-bold-weight:600;line-height:1.3em; }.tcb-style-wrap h3{ font-family:"Open Sans";--g-regular-weight:400;font-weight:var(--g-bold-weight, bold);color:var(--tcb-skin-color-4);--tcb-applied-color:var$(--tcb-skin-color-4);--g-bold-weight:600;padding-top:0px;margin-top:0px;padding-bottom:10px;margin-bottom:0px;font-size:26px;line-height:1.4em; }.tcb-style-wrap h4{ font-family:"Open Sans";--g-regular-weight:400;font-weight:var(--g-bold-weight, bold);color:var(--tcb-skin-color-4);--tcb-applied-color:var$(--tcb-skin-color-4);--g-bold-weight:600;padding-top:0px;margin-top:0px;padding-bottom:10px;margin-bottom:0px;font-size:20px;line-height:1.4em; }.tcb-style-wrap h5{ font-family:"Open Sans";--g-regular-weight:400;font-weight:var(--g-bold-weight, bold);color:var(--tcb-skin-color-4);--tcb-applied-color:var$(--tcb-skin-color-4);--g-bold-weight:600;padding-top:0px;margin-top:0px;padding-bottom:10px;margin-bottom:0px;line-height:1.4em; }.tcb-style-wrap h6{ font-family:"Open Sans";--g-regular-weight:400;font-weight:var(--g-bold-weight, bold);color:var(--tcb-skin-color-4);--tcb-applied-color:var$(--tcb-skin-color-4);--g-bold-weight:600;padding-top:0px;margin-top:0px;padding-bottom:10px;margin-bottom:0px;line-height:1.6em; }.tcb-style-wrap p{ color:var(--tcb-color-13);--tcb-applied-color:var$(--tcb-color-13);font-family:"Open Sans";--g-regular-weight:400;--g-bold-weight:700;font-weight:var(--g-regular-weight, normal);font-size:15px;padding-top:15px;margin-top:0px;padding-bottom:15px;margin-bottom:0px; }:not(.inc) .thrv_text_element a, :not(.inc) .tcb-styled-list a, :not(.inc) .tcb-numbered-list a, .tcb-post-content p a, .tcb-post-content li a, .tcb-post-content blockquote a, .tcb-post-content pre a{ --tcb-applied-color:var$(--tcb-skin-color-5);font-family:inherit;--g-regular-weight:400;--g-bold-weight:800;font-weight:var(--g-regular-weight, normal);font-size:inherit;text-decoration-line:none;--eff:none;text-decoration-color:var(--eff-color, currentColor);color:var(--tcb-color-14); }:not(.inc) .thrv_text_element a:hover, :not(.inc) .tcb-styled-list a:hover, :not(.inc) .tcb-numbered-list a:hover, .tcb-post-content p a:hover, .tcb-post-content li a:hover, .tcb-post-content blockquote a:hover, .tcb-post-content pre a:hover { text-decoration-line:none;--eff:none;text-decoration-color:var(--eff-color, currentColor);color:var(--tcb-skin-color-3); }.tcb-style-wrap ul, .tcb-style-wrap ol{ color:var(--tcb-skin-color-5);--tcb-applied-color:var$(--tcb-skin-color-5);font-family:Muli;--g-regular-weight:400;--g-bold-weight:800;font-weight:var(--g-regular-weight, normal);font-size:18px;margin-top:0px;margin-bottom:0px;padding-top:15px;padding-bottom:15px;padding-left:0px; }.tcb-style-wrap li{ color:var(--tcb-color-5);--tcb-applied-color:var$(--tcb-color-5);font-family:"Open Sans";--g-regular-weight:400;--g-bold-weight:700;font-weight:var(--g-regular-weight, normal);font-size:15px;line-height:1.4em;padding-bottom:20px;margin-bottom:0px; }.tcb-style-wrap blockquote{ color:var(--tcb-color-5);--tcb-applied-color:var$(--tcb-color-5);font-family:"Open Sans";--g-regular-weight:400;--g-bold-weight:800;font-weight:var(--g-regular-weight, normal);font-size:15px;padding-top:15px;padding-bottom:15px;border-left:2px solid var(--tcb-skin-color-1); }body{ color:var(--tcb-color-5);--tcb-applied-color:var$(--tcb-color-5);font-family:"Open Sans";--g-regular-weight:400;--g-bold-weight:700;font-weight:var(--g-regular-weight, normal);font-size:15px; }.tcb-style-wrap pre{ color:var(--tcb-color-12);--tcb-applied-color:var$(--tcb-color-12);font-size:15px; } }@media (max-width: 1023px) { .tcb-style-wrap h1{ font-size:36px;padding-bottom:8px;margin-bottom:0px; }.tcb-style-wrap h3{ padding-bottom:8px;margin-bottom:0px;font-size:22px; }.tcb-style-wrap h4{ padding-bottom:8px;margin-bottom:0px; }.tcb-style-wrap h5{ padding-bottom:8px;margin-bottom:0px; }.tcb-style-wrap h6{ padding-bottom:8px;margin-bottom:0px; }.tcb-style-wrap p{ padding-bottom:10px;margin-bottom:0px;padding-top:10px;margin-top:0px; }.tcb-style-wrap ul, .tcb-style-wrap ol{ padding-top:10px;padding-bottom:10px; } }@media (max-width: 767px) { .tcb-style-wrap h1{ font-size:28px; }.tcb-style-wrap h3{ font-size:22px; }.tcb-style-wrap h4{ font-size:20px; }.tcb-style-wrap h6{ padding-bottom:8px;margin-bottom:0px; }.tcb-style-wrap p{ padding-bottom:5px;margin-bottom:0px;padding-top:5px;margin-top:0px;font-size:15px; }.tcb-style-wrap ul, .tcb-style-wrap ol{ padding-left:0px;padding-bottom:5px;padding-top:5px; }.tcb-style-wrap li{ font-size:15px; }body{ font-size:15px; } }</style><link rel="icon" href="https://www.dataquest.io/wp-content/uploads/2020/04/favicon-32x32-1.png" sizes="32x32" />
<link rel="icon" href="https://www.dataquest.io/wp-content/uploads/2020/04/favicon-32x32-1.png" sizes="192x192" />
<link rel="apple-touch-icon" href="https://www.dataquest.io/wp-content/uploads/2020/04/favicon-32x32-1.png" />
<meta name="msapplication-TileImage" content="https://www.dataquest.io/wp-content/uploads/2020/04/favicon-32x32-1.png" />
<style type="text/css" class="tve_global_style"></style>			<style type="text/css"
				   class="tve_custom_style">@media (min-width: 300px){[data-css="tve-u-16aa2e420ed"] p { margin: 0px !important; padding: 0px !important; }[data-css="tve-u-16aa2e420ed"] h1 { margin: 0px !important; padding: 0px !important; }[data-css="tve-u-16aa2e420ef"] strong { font-weight: 700; }[data-css="tve-u-16aa2e420ef"] { display: block; max-width: 250px; margin-bottom: -42px !important; }[data-css="tve-u-16aa2e420ef"] .tcb-button-link { border: 2px solid rgb(77, 127, 223); background-image: none !important; background-color: rgb(97, 209, 153) !important; padding: 20px !important; }[data-css="tve-u-16aa2e420f2"] { background-image: none !important; margin-top: 0px !important; margin-bottom: 60px !important; }[data-css="tve-u-16aa2e420ed"] { min-height: 97px; }[data-css="tve-u-16aa2e420f3"] { background-image: none !important; background-color: rgb(255, 255, 255) !important; border: 2px solid rgb(77, 127, 223)  !important; }[data-css="tve-u-16aa2e420f4"] { padding: 43px 43px 10px !important; margin-top: 40px !important; margin-bottom: 80px !important; }[data-css="tve-u-16aa2e420f6"] { background-image: none !important; margin-top: -48px !important; padding-bottom: 0px !important; margin-bottom: 30px !important; }}@media (max-width: 767px){[data-css="tve-u-16aa2e420f6"] { background-image: none !important; }}</style>		<style type="text/css" id="wp-custom-css">
			body   {font-family: "Open Sans", sans-serif; color: #232a3d;}

.tcb-button-text { font-weight: 600; letter-spacing: 0px;}

h1, h2 {font-family: "Industry", sans-serif!important;  font-weight: bold!important;}

.post-template-default h2 {font-family: "Open Sans", sans-serif!important; padding-top: 30px;}

.path-box h3 {font-size: 17px!important; font-weight: 900; margin: 0px 0px 5px 0px; line-height: 1em}

.path-box p {line-height: 1.6em; font-size: 14px;}

.path-box a {text-decoration: none!important; }

.path-box-button {height: 50px;}

a .path-box-button-text p  {text-transform: uppercase!important; letter-spacing: 1px; color: #ffffff!important; margin-top: 0px; padding-top: 4px;}



:not(pre) > code[class*="language-"], pre[class*="language-"] {background-color: #000000; border-radius: 3px;}

p code {
background-color: #e5eff5;
padding: 1px 3px;
border-radius: 3px;
font-size: .95em;
color: #000000;}

pre {background: #000000; margin-top: -10px;}
pre code {font-size: 13px; padding: 0px; color: #ffffff; background: #000000;}

code[class*="language-"], pre[class*="language-"] {font-size: 15px; margin-bottom: 20px; margin-top: -10px;}


.single-post table {width: 85%; margin: 30px auto; }

.single-post table th {background-color: #f4f8fb; /*text-transform: lowercase;*/ color: #232a3d!important;}

.single-post table td, article table th {font-size: 15px; border: 1px solid #e3ecf3; font-family: ingra, sans-serif; }


div table{
    overflow-x:auto;
}





.tvo-set12-template.tve_teal .tvo-testimonial-quote {color: #4d7fdf!important;}

.tvo-set12-template.tve_teal .tvo-testimonial-name {color: #4d7fdf!important;}


.directory-coursebox p {margin-top: -20px; margin-bottom: 14px!important; }

.directory-coursebox div.thrv_wrapper.thrv-button {padding-top: 16px;}

.directory-course-label p {margin: 4px 0px; display: none;}


@media only screen and (max-width: 769px) {
	.directory-coursebox .tcb--cols--2, .directory-coursebox .tcb--cols--2 p { text-align: center!important; }}

@media only screen and (max-width: 769px) {
	.directory-coursebox .tcb--cols--2 .directory-course-label p { margin-top: -20px;}}

@media only screen and (max-width: 769px) {
	.directory-coursebox .tcb--cols--2 .thrv-button a { margin-top: -20px;}}

.directory-coursebox .tcb-button-texts {padding: 0px 20px;}

@media only screen and (max-width: 767px){.thrv_wrapper.thrv-columns.directory-course-columns {margin-top: -40px;}}

.path-box-button-text p {font-family: industry!important; font-weight: 700;}

.course-box-course-info p {font-family: industry!important; font-weight: 700; font-size: 14px}

.course-box h2 {font-size: 22px;}

.course-box p {font-size: 13px;}

.course-box a {text-decoration: none!important;}

.scn.aut {
    box-shadow: none;
    background: #ffffff;
    border-radius: 5px;
    border: 2px solid #d4d4d4;
    margin-bottom: 10px;
}


.tvo-set12-template.tvo-testimonials-display-single .tvo-testimonial-quote {font-family: industry!important; font-size: 168px!important; margin-left: -10px;}

.tvo-set12-template.tvo-testimonials-display-single p {margin-top: -40px;}

.tvo-testimonial-image-cover .tvo-testimonial-real-border {display: block!important;}


.tvo-testimonial-info svg {display: none; }


.tvo-set12-template.tvo-testimonials-display-single .tvo-testimonial-quote:before, .tvo-set12-template.tvo-testimonials-display-single .tvo-testimonial-quote:after {display: none;}

@media only screen and (min-width: 768px) {
	.why-p, .mission-two-columns {column-count: 2; column-gap: 40px; }}

.why-p p, .why-p ul li {font-size: 15px; margin-bottom: 20px!important; font-weight: 300;}

.gmt-edd-slack-alert-success {border: 1px solid #000000; padding: 20px; text-align: center;}

.sub-picker .tve_scT .tve_scTC {background: transparent; border: 0px;}

.sub-picker .tve_scT>ul li {border: 1px solid #cccccc; text-transform: capitalize; }

.sub-picker .tve_scT>ul li.annual {border-radius: 5px;}

.sub-picker .tve_scT>ul li.monthly {border-radius: 5px; margin-bottom: 5px!important;}

.thrv_wrapper.thrv-tabbed-content div.tve_scT>ul li span {width: 70px; font-size: 12px; font-family: industry; letter-spacing: 1px; line-height: 1em; }

.sub-picker ul.mon-ann {display: block; margin: auto auto!important; width: 200px; }

.sub-picker ul li.annual, .sub-picker ul li.monthly {padding: 10px!important; }

.thrv-pricing-table {margin-top: -40px}

.thrv_wrapper.thrv-tabbed-content div.tve_scT>ul li:hover {color: #232a3d; }

.main-testimonial .tvo-testimonials-display.tvo-set12-template .tvo-testimonial-display-item p {font-size: 20px!important;}

.mission-ppp-horizontal {display: none;}

.mission-outline p {font-size: 14px;  line-height: 1.6em; padding-top: 8px;}

.mission-outline br {line-height: 3em}

.mission-outline {column-count: 1; column-gap: 10px; }

.ctb {border: none;}
.ctb h5  {display: none;}



.lrp p {font-family: industry, sans-serif!important; font-size: 18px!important; color: #232a3d!important;}

textarea:not(.editor-post-title__input) {background: #ffffff; box-shadow: none; border: 2px solid #e9e9e9; }

.lrp input[type="submit"] {
    background: #39CF90;
    border-radius: 5px;
	border: 0px;
    color: #ffffff;
    font-size: 1em;
    font-family: industry, sans-serif;
    margin: 20px 0;
    padding: 12px 36px 12px 36px;
    text-align: center;
}

.lrp input[type="text"] {
    box-sizing: border;
    border-radius: 5px;
	box-shadow: none;
    background: #ffffff;
    border: 1px solid #e9e9e9;
    color: #232a3d;
    float: left;
    font-size: 1em;
    font-weight: 700;
    font-family: ingra, sans-serif;
    height: 32px;
    margin: 0% 4.2% 3% 0%;
    padding: 0% 0% 0% 2%;
    width: 30.3%;
}

.lrp textarea::-webkit-input-placeholder {
   color:transparent;
}

.lrp textarea:-moz-placeholder { /* Firefox 18- */
   color:transparent; 
}

.lrp textarea::-moz-placeholder {  /* Firefox 19+ */
   color:transparent; 
}

.lrp textarea:-ms-input-placeholder {  
   color:transparent;
}





.thrv_countdown_timer .tve_t_part .t-digits {color: #ffffff!important; margin: 0px!important;}

.tve-leads-ribbon .thrv_countdown_timer .tve_t_part .t-digits {color: #ffffff!important; margin: 0px!important; font-size: 70%!important; margin-bottom: -8px!important; letter-spacing: 1;}




.course-header p strong, .course-header .tcb-plain-text, .mission-header p strong, .mission-header .tcb-plain-text {font-size: 12px!important; letter-spacing: 3px;
    color: rgb(77, 127, 223) !important;}

.mission-header .tcb-plain-text {margin-bottom: 10px;}



.tve_flt .tve_contents_table .tve_ct_title {font-family: industry; text-align: left; display: block; margin: 0px ; padding: 0px 0px 10px; font-weight: bold; font-size: 24px;}

.tve-leads-ribbon.tve-leads-triggered {
    z-index: 100;
}

header#thrive-header {z-index: 10;}		</style>
		</head>

<body class="post-template-default single single-post postid-2024 single-format-standard locale-en-us tve-theme-23111 thrive-layout-23090">
<div id="wrapper" class="tcb-style-wrap"><header id="thrive-header" class="thrv_wrapper thrv_symbol thrv_header thrv_symbol_23138" role="banner"><div class="thrive-shortcode-html thrive-symbol-shortcode" data-symbol-id="23138"><style class='tve-symbol-custom-style'>@media (min-width: 300px){.thrv_symbol_23138 [data-css="tve-u-16eff063edc"] { padding: 0px 20px !important; }.thrv_symbol_23138 [data-css="tve-u-16eff064345"] { margin: 0px !important; }.thrv_symbol_23138 [data-css="tve-u-16eff063edc"] > .tcb-flex-col > .tcb-col { justify-content: center; }.thrv_symbol_23138 [data-css="tve-u-16eff067282"] { max-width: 17.3%; }.thrv_symbol_23138 [data-css="tve-u-16eff067395"] { max-width: 60.7%; }.thrv_symbol_23138 .symbol-section-out { box-shadow: rgba(0, 0, 0, 0.1) 0px 4px 12px 0px; background-color: rgb(255, 255, 255)  !important; }.thrv_symbol_23138 [data-css="tve-u-1705d5b97bd"] { float: left; position: relative; z-index: 10 !important; margin-top: 0px !important; margin-bottom: 0px !important; }.thrv_symbol_23138 .thrv_widget_menu #m-1705d5b97ba > li:not(.tcb-excluded-from-group-item) { --tve-font-size:13px; --tve-color:rgb(137, 138, 142); --tve-font-weight:var(--g-regular-weight, normal) ; margin-left: 0px !important; margin-right: 0px !important; padding-left: 15px !important; padding-right: 15px !important; }:not(#tve) .thrv_symbol_23138 .thrv_widget_menu #m-1705d5b97ba > li:not(.tcb-excluded-from-group-item):hover > a { color: var(--tve-color, var(--tcb-skin-color-4))  !important; --tcb-applied-color:var$(--tcb-skin-color-4)  !important; }:not(#tve) .thrv_symbol_23138 .thrv_widget_menu #m-1705d5b97ba > li:not(.tcb-excluded-from-group-item):hover { --tve-color:var(--tcb-skin-color-4)  !important; }:not(#tve) .thrv_symbol_23138 .thrv_widget_menu #m-1705d5b97ba > li:not(.tcb-excluded-from-group-item) > a { font-weight: var(--tve-font-weight, var(--g-regular-weight, normal) )  !important; }.thrv_symbol_23138 [data-css="tve-u-1705d61006e"]::after { clear: both; }.thrv_symbol_23138 [data-css="tve-u-1705d6234a9"] { max-width: 9.4%; }.thrv_symbol_23138 [data-css="tve-u-1705d64bb4b"]::after { clear: both; }.thrv_symbol_23138 [data-css="tve-u-17076e1c331"] { width: 150px; margin-top: 0px !important; margin-bottom: 0px !important; }.thrv_symbol_23138 .symbol-section-in { padding: 15px 20px !important; }.thrv_symbol_23138 [data-css="tve-u-171451bc8cd"] { --tcb-local-color-62516:rgb(57, 207, 144); --tve-font-size:16px; display: block; max-width: 115px; width: 115px; margin: 0px !important; }.thrv_symbol_23138 [data-css="tve-u-171451bc8cd"] .tcb-button-link { letter-spacing: 2px; border-radius: 5px; overflow: hidden; background-image: none !important; padding: 5px !important; background-color: rgb(57, 207, 144)  !important; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-171451bc8cd"] .tcb-button-link { font-size: var(--tve-font-size, 16px)  !important; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145eed6a0"] p, :not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145eed6a0"] li, :not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145eed6a0"] blockquote, :not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145eed6a0"] address, :not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145eed6a0"] .tcb-plain-text, :not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145eed6a0"] label { font-size: var(--tve-font-size, 14px); }.thrv_symbol_23138 [data-css="tve-u-17145eed6a0"] { --tve-font-size:14px; }.thrv_symbol_23138 .thrv_widget_menu #m-1705d5b97ba > li:not(.tcb-excluded-from-group-item) > a { --tcb-applied-color:var$(--tcb-color-8); font-size: var(--tve-font-size, 13px)  !important; color: var(--tve-color, rgb(137, 138, 142))  !important; }.thrv_symbol_23138 [data-css="tve-u-17145ef5aa8"] { --tcb-local-color-62516:rgb(255, 255, 255); --tve-font-size:16px; --tve-color:rgb(73, 118, 234); display: block; max-width: 95px; width: 95px; min-width: 107px; margin: 0px !important; }.thrv_symbol_23138 [data-css="tve-u-17145ef5aa8"] .tcb-button-link { letter-spacing: 2px; border-radius: 5px; overflow: hidden; border: none; background-image: none !important; padding: 5px 20px 5px 5px !important; background-color: rgb(255, 255, 255)  !important; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145ef5aa8"] .tcb-button-link { font-size: var(--tve-font-size, 16px)  !important; }.thrv_symbol_23138 [data-css="tve-u-17145ef781f"] { max-width: 12.6%; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145f19d67"] { font-size: 13px !important; color: rgb(137, 138, 142)  !important; --tcb-applied-color:var$(--tcb-color-8)  !important; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145f1aef7"] { font-size: 14px !important; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145ef5aa8"] .tcb-button-link span { color: var(--tve-color, rgb(73, 118, 234)); --tcb-applied-color:var$(--tcb-color-14); }.thrv_symbol_23138 { z-index: 9999; top: 0px; position: absolute !important; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-171451bc8cd"]:hover .tcb-button-link { background-color: rgb(55, 186, 131) !important; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145ef5aa8"]:hover .tcb-button-link span { color: var(--tve-color, rgb(137, 138, 142))  !important; --tcb-applied-color:var$(--tcb-color-8)  !important; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145ef5aa8"]:hover { --tve-color:rgb(137, 138, 142)  !important; }.thrv_symbol_23138 [data-css="tve-u-1715bd6bfbc"] { font-size: 13px; width: 13px; height: 13px; margin-right: -10px !important; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-171451bc8cd"] { --tcb-local-color-62516:rgb(73, 118, 234); }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-1715bd6bfbc"] > :first-child { color: rgb(137, 138, 142); }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-1715bd6bfbc"]:hover > :first-child { color: rgb(35, 42, 61)  !important; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-17145f19d67"]:hover { color: rgb(35, 42, 61)  !important; --tcb-applied-color:var$(--tcb-color-10)  !important; font-weight: var(--g-regular-weight, normal)  !important; }.thrv_symbol_23138 [data-css="tve-u-17145f19d67"] { font-weight: var(--g-regular-weight, normal)  !important; }}@media (max-width: 1023px){.thrv_symbol_23138 [data-css="tve-u-16eff063edc"] { padding: 0px !important; flex-flow: row wrap !important; }.thrv_symbol_23138 [data-css="tve-u-16eff063edc"] .tcb-flex-col { flex-basis: 203px !important; }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-1705d5b97bd"] .tve-m-trigger .thrv_icon > svg { color: var(--tcb-skin-color-5); }.thrv_symbol_23138 [data-css="tve-u-1705d5b97bd"] .tve-m-trigger .thrv_icon { font-size: 33px; width: 33px; height: 33px; }.thrv_symbol_23138 [data-css="tve-u-1705d5b97bd"] { float: none; margin-left: auto !important; margin-right: auto !important; }.thrv_symbol_23138 .symbol-section-in { padding: 8px 40px !important; }.thrv_symbol_23138 [data-css="tve-u-1705d5b97bd"] .tve_w_menu { padding: 20px 40px !important; }.thrv_symbol_23138 .thrv_widget_menu #m-1705d5b97ba > li:not(.tcb-excluded-from-group-item) { --tve-line-height:1.65em; padding-left: 0px !important; padding-right: 0px !important; }.thrv_symbol_23138 .thrv_widget_menu #m-1705d5b97ba > li:not(.tcb-excluded-from-group-item) > a { line-height: var(--tve-line-height, 1.65em)  !important; }.thrv_symbol_23138 [data-css="tve-u-17076e1c331"] { padding-left: 20px !important; }.thrv_symbol_23138 [data-css="tve-u-1716018dbea"] { padding-right: 20px !important; }.thrv_symbol_23138 [data-css="tve-u-171451bc8cd"] { display: block; max-width: 115px; width: 115px; float: none; z-index: 3; position: relative; min-width: 0px !important; margin-left: auto !important; margin-right: auto !important; }}@media (max-width: 767px){.thrv_symbol_23138 [data-css="tve-u-16eff063edc"] { padding: 0px !important; flex-flow: row wrap !important; }.thrv_symbol_23138 [data-css="tve-u-16eff063edc"] .tcb-flex-col { flex-basis: 203px !important; }.thrv_symbol_23138 [data-css="tve-u-16eff063edc"] > .tcb-flex-col { flex-basis: 85px !important; }.thrv_symbol_23138 [data-css="tve-u-1705d5b97bd"] { float: right; position: relative; z-index: 10 !important; }.thrv_symbol_23138 [data-css="tve-u-1705d61006e"]::after { clear: both; }.thrv_symbol_23138 [data-css="tve-u-1705d5b97bd"] .tve-m-trigger .thrv_icon { font-size: 29px; width: 29px; height: 29px; }.thrv_symbol_23138 .symbol-section-in { padding: 4px 10px 4px 20px !important; }.thrv_symbol_23138 [data-css="tve-u-1705d5b97bd"] .tve_w_menu { border: 1px solid rgb(73, 118, 234); }:not(#tve) .thrv_symbol_23138 [data-css="tve-u-1705d5b97bd"] .tve-m-trigger .thrv_icon > svg { color: rgb(73, 118, 234); }.thrv_symbol_23138 [data-css="tve-u-171451bc8cd"] .tcb-button-link { padding-left: 5px !important; padding-right: 5px !important; }.thrv_symbol_23138 [data-css="tve-u-171451bc8cd"] { display: block; max-width: 80px; width: 80px; }}</style><div class="symbol-section-out" style=""></div> <div class="symbol-section-in" style=""><div class="thrv_wrapper thrv-columns tcb-window-width" data-css="tve-u-16eff064345" style="width: 1111px; left: -35.5px;"><div class="tcb-flex-row tcb-medium-wrap v-2 m-edit tcb-mobile-wrap tcb--cols--4 tcb-resized" style="padding-bottom: 0px !important;" data-css="tve-u-16eff063edc"><div class="tcb-flex-col" data-css="tve-u-16eff067282" style=""><div class="tcb-col"><a class="tcb-logo thrv_wrapper" href="https://www.dataquest.io" data-css="tve-u-17076e1c331" class=""><picture><source data-srcset="https://www.dataquest.io/wp-content/uploads/2020/04/dataquest_logo_nav6f10119225a06c6415932df9b22c93da-300x64.png 300w, https://www.dataquest.io/wp-content/uploads/2020/04/dataquest_logo_nav6f10119225a06c6415932df9b22c93da.png 336w" media="(min-width:1024px)"></source><noscript><img src="https://www.dataquest.io/wp-content/uploads/2020/04/dataquest_logo_nav6f10119225a06c6415932df9b22c93da.png" height="72" width="336" alt="" loading="lazy"></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20336%2072%22%3E%3C/svg%3E' data-src="https://www.dataquest.io/wp-content/uploads/2020/04/dataquest_logo_nav6f10119225a06c6415932df9b22c93da.png" height="72" width="336" alt="" loading="lazy"></img></picture></a></div></div><div class="tcb-flex-col" data-css="tve-u-16eff067395" style=""><div class="tcb-col" style="" data-css="tve-u-17145eed6a0"><div class="tcb-clear" data-css="tve-u-1705d61006e"><div class="thrv_wrapper thrv_widget_menu tve-custom-menu-upgrade tve-mobile-dropdown tve-menu-template-light-tmp-first da-fold tve-regular" data-tve-switch-icon="tablet,mobile" data-css="tve-u-1705d5b97bd" style=""><div class="thrive-shortcode-config" style="display:none"></div><div class="thrive-shortcode-html thrive-shortcode-html-editable tve_clearfix" ><a class="tve-m-trigger t_tve_horizontal" href="javascript:void(0)"><div class="thrv_wrapper thrv_icon tcb-icon-open" data-not-changeable="true"><svg class="tcb-icon" viewBox="0 0 32 32" data-name="align-justify"><g><rect x="8" y="10.1" class="st0" width="16" height="2"/><rect x="8" y="15" class="st0" width="16" height="2"/><rect x="8" y="19.9" class="st0" width="16" height="2"/><path class="st0" d="M0,0v32h32V0H0z M30,30H2V2h28V30z"/></g></svg></div><div class="thrv_wrapper thrv_icon tcb-icon-close" data-not-changeable="true"><svg class="tcb-icon" viewBox="0 0 32 32" data-name="close"><g><path class="st0" d="M0,0v32h32V0H0z M30,30H2V2h28V30z"/><polygon class="st0" points="11.1,22.4 16,17.4 21,22.4 22.4,21 17.4,16 22.4,11.1 21,9.6 16,14.6 11.1,9.6 9.6,11.1 14.6,16 9.6,21 "/></g></svg></div></a><ul id="m-1705d5b97ba" class="tve_w_menu tve_horizontal "><li id="menu-item-20226" class="menu-item menu-item-type-post_type menu-item-object-page menu-item-20226 lvl-0" data-id="20226"><a href="https://www.dataquest.io/directory/"><span class="tve-disabled-text-inner">Courses</span></a></li>
<li id="menu-item-23264" class="menu-item menu-item-type-taxonomy menu-item-object-category menu-item-23264 lvl-0" data-id="23264"><a href="https://www.dataquest.io/blog/topics/student-stories/"><span class="tve-disabled-text-inner">Student Stories</span></a></li>
<li id="menu-item-20220" class="menu-item menu-item-type-post_type menu-item-object-page current_page_parent menu-item-20220 lvl-0" data-id="20220"><a href="https://www.dataquest.io/blog/"><span class="tve-disabled-text-inner">Blog</span></a></li>
<li id="menu-item-23566" class="menu-item menu-item-type-post_type menu-item-object-page menu-item-23566 lvl-0" data-id="23566"><a href="https://www.dataquest.io/were-hiring/"><span class="tve-disabled-text-inner">We’re Hiring</span></a></li>
<li id="menu-item-22382" class="menu-item menu-item-type-custom menu-item-object-custom menu-item-22382 lvl-0" data-id="22382"><a href="https://www.dataquest.io/for-business/"><span class="tve-disabled-text-inner">For Business</span></a></li>
</ul></div> </div></div></div></div><div class="tcb-flex-col" data-css="tve-u-17145ef781f" style=""><div class="tcb-col" style="" data-css="tve-u-1716018dbea"><div class="thrv_wrapper thrv-button thrv-button-v2 tcb-local-vars-root" data-css="tve-u-171451bc8cd" data-button-size="s" style="" data-tcb_hover_state_parent=""> <div class="thrive-colors-palette-config" style="display: none !important">__CONFIG_colors_palette__{"active_palette":0,"config":{"colors":{"62516":{"name":"Main Accent","parent":-1}},"gradients":[]},"palettes":[{"name":"Default Palette","value":{"colors":{"62516":{"val":"var(--tcb-color-15)","hsl":{"h":154,"s":0.61,"l":0.01}}},"gradients":[]}}]}__CONFIG_colors_palette__</div> <a href="https://app.dataquest.io/signup" class="tcb-button-link" style=""> <span class="tcb-button-texts"><span class="tcb-button-text thrv-inline-text" style="" data-css="tve-u-17145f1aef7">Start Free</span></span> </a> </div></div></div><div class="tcb-flex-col" data-css="tve-u-1705d6234a9" style=""><div class="tcb-col" data-css="tve-u-1705d64bb4b"><div class="thrv_wrapper thrv-button thrv-button-v2 tcb-local-vars-root tcb-tablet-hidden tcb-mobile-hidden tcb-with-icon" data-css="tve-u-17145ef5aa8" data-button-size="s" style="" data-class=" tcb-tablet-hidden tcb-mobile-hidden" data-tcb_hover_state_parent=""> <div class="thrive-colors-palette-config" style="display: none !important">__CONFIG_colors_palette__{"active_palette":0,"config":{"colors":{"62516":{"name":"Main Accent","parent":-1}},"gradients":[]},"palettes":[{"name":"Default Palette","value":{"colors":{"62516":{"val":"var(--tcb-color-12)","hsl":{"h":0,"s":0.01,"l":0.01}}},"gradients":[]}}]}__CONFIG_colors_palette__</div> <a href="https://app.dataquest.io/dashboard" class="tcb-button-link" style="padding-left: 5px !important;"><span class="tcb-button-icon" style=""> <div class="thrv_wrapper thrv_icon tve_no_drag tve_no_icons tcb-icon-inherit-style tcb-icon-display" style="" data-css="tve-u-1715bd6bfbc" data-tcb_hover_state_parent=""><svg class="tcb-icon" viewBox="0 0 512 512" data-id="icon-user-alt-regular" data-name="" style=""> <path d="M384 336c-40.6 0-47.6-1.5-72.2 6.8-17.5 5.9-36.3 9.2-55.8 9.2s-38.3-3.3-55.8-9.2c-24.6-8.3-31.5-6.8-72.2-6.8C57.3 336 0 393.3 0 464v16c0 17.7 14.3 32 32 32h448c17.7 0 32-14.3 32-32v-16c0-70.7-57.3-128-128-128zm80 128H48c0-21.4 8.3-41.5 23.4-56.6C86.5 392.3 106.6 384 128 384c41.1 0 41-1.1 56.8 4.2 23 7.8 47 11.8 71.2 11.8 24.2 0 48.2-4 71.2-11.8 15.8-5.4 15.7-4.2 56.8-4.2 44.1 0 80 35.9 80 80zM256 320c88.4 0 160-71.6 160-160S344.4 0 256 0 96 71.6 96 160s71.6 160 160 160zm0-272c61.8 0 112 50.2 112 112s-50.2 112-112 112-112-50.2-112-112S194.2 48 256 48z"></path> </svg></div> </span> <span class="tcb-button-texts" style=""><span class="tcb-button-text thrv-inline-text" style="" data-css="tve-u-17145f19d67" data-tcb_hover_state_parent="">Dashboard</span></span> </a> </div></div></div></div></div></div><style type="text/css" class="tve_global_style"></style></div></header><div id="theme-top-section" class="theme-section top-section"><div class="section-background"></div><div class="section-content"><div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-1705cdb4152"><h1 class="">​​​​<span class="thrive-shortcode-content" data-attr-link="0" data-attr-rel="0" data-attr-target="0" data-css="tve-u-1707753ab8f" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_title" data-shortcode-name="Post title">Jupyter Notebook for Beginners: A Tutorial</span></h1></div></div></div><div id="content">
<div class="main-container thrv_wrapper">
<aside id="theme-sidebar-section" class="theme-section sidebar-section thrive-section-23262" role="complementary"><div class="section-background"></div><div class="section-content">
<div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad">
	<div class="tve-content-box-background" style="" data-css="tve-u-1714ad3b039"></div>
	<div class="tve-cb"><div class="thrv_wrapper thrv_icon tcb-icon-display" data-css="tve-u-1714ad51f42" style=""><svg class="tcb-icon" viewBox="0 0 576 512" data-id="icon-code-regular" data-name="" style="">
            <path d="M234.8 511.7L196 500.4c-4.2-1.2-6.7-5.7-5.5-9.9L331.3 5.8c1.2-4.2 5.7-6.7 9.9-5.5L380 11.6c4.2 1.2 6.7 5.7 5.5 9.9L244.7 506.2c-1.2 4.3-5.6 6.7-9.9 5.5zm-83.2-121.1l27.2-29c3.1-3.3 2.8-8.5-.5-11.5L72.2 256l106.1-94.1c3.4-3 3.6-8.2.5-11.5l-27.2-29c-3-3.2-8.1-3.4-11.3-.4L2.5 250.2c-3.4 3.2-3.4 8.5 0 11.7L140.3 391c3.2 3 8.2 2.8 11.3-.4zm284.1.4l137.7-129.1c3.4-3.2 3.4-8.5 0-11.7L435.7 121c-3.2-3-8.3-2.9-11.3.4l-27.2 29c-3.1 3.3-2.8 8.5.5 11.5L503.8 256l-106.1 94.1c-3.4 3-3.6 8.2-.5 11.5l27.2 29c3.1 3.2 8.1 3.4 11.3.4z"></path>
        </svg></div><div class="thrv_wrapper thrv_text_element"><h3 class="" style="text-align: center;">Learn by <span data-css="tve-u-170a20da55e" style="display: inline-block; text-decoration: line-through;">watching videos</span>&nbsp;<span data-css="tve-u-170a20da561">coding!</span></h3></div><div class="thrv_wrapper thrv-button thrv-button-v2 tcb-local-vars-root tcb-global-button-k8mi4xps" data-button-size="s" data-css="tve-u-1714acfc836" data-button-style="tcb-global-button-k8mi4xps" style="display: block; max-width: 100%; width: 100%;" data-tcb_hover_state_parent=""><div class="thrive-colors-palette-config" style="display: none !important">__CONFIG_colors_palette__{"active_palette":0,"config":{"colors":{"493ef":{"name":"Main Accent","parent":-1}},"gradients":[]},"palettes":[{"name":"Default Palette","value":{"colors":{"493ef":{"val":"var(--tcb-color-15)","hsl":{"h":154,"s":0.61,"l":0.01}}},"gradients":[]},"original":{"colors":{"493ef":{"val":"rgb(19, 114, 211)","hsl":{"h":210,"s":0.83,"l":0.45}}},"gradients":[]}}]}__CONFIG_colors_palette__</div>
		<a href="https://app.dataquest.io/signup" class="tcb-button-link" style="">
		<span class="tcb-button-texts"><span class="tcb-button-text thrv-inline-text tcb-global-button-k8mi4xps-prtext">Try it now &gt;&gt;</span></span>
	</a>
</div></div>
</div><div id="" class="thrv_wrapper thrv-search-form " data-css="" data-tcb-events="" data-ct-name="Default Template" data-ct="search_form-56234">
	<form role="search" method="get" action="https://www.dataquest.io/?s=post">
		<div class="thrv-sf-submit" data-button-layout="icon" data-css="tve-u-171642ab6db">
			<button type="submit">
				<span class="tcb-sf-button-icon">
					<span class="thrv_wrapper thrv_icon tve_no_drag tve_no_icons tcb-icon-inherit-style tcb-icon-display" data-css=""><svg class="tcb-icon" viewBox="0 0 512 512" data-id="icon-search-regular"><path d="M508.5 468.9L387.1 347.5c-2.3-2.3-5.3-3.5-8.5-3.5h-13.2c31.5-36.5 50.6-84 50.6-136C416 93.1 322.9 0 208 0S0 93.1 0 208s93.1 208 208 208c52 0 99.5-19.1 136-50.6v13.2c0 3.2 1.3 6.2 3.5 8.5l121.4 121.4c4.7 4.7 12.3 4.7 17 0l22.6-22.6c4.7-4.7 4.7-12.3 0-17zM208 368c-88.4 0-160-71.6-160-160S119.6 48 208 48s160 71.6 160 160-71.6 160-160 160z"></path></svg></span>
				</span>
				<span class="tve_btn_txt">Search</span>
			</button>
		</div>
		<div class="thrv-sf-input" data-css="tve-u-171642ab6d6">
			<input type="search" placeholder="Search" value="" name="s"/>
		</div>
					<input type="hidden" class="tcb_sf_post_type" name="tcb_sf_post_type[]" value="post" data-label="Post"/>
					<input type="hidden" class="tcb_sf_post_type" name="tcb_sf_post_type[]" value="page" data-label="Page"/>
			</form>
</div><div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-17058d3d5d0"><p data-css="tve-u-17058d2f3b8">categories</p></div><div class="thrive-shortcode thrive-dynamic-list thrv_wrapper" data-type="categories" data-css="tve-u-17058d6fd4a"><ul class="theme-dynamic-list">
			<li class="thrive-dynamic-styled-list-item tve_no_icons" data-selector=".thrive-dynamic-styled-list-item">
			<div class="tcb-styled-list-icon">
							</div>
			<div class="thrive-dynamic-styled-list-text" data-selector=".thrive-dynamic-styled-list-text a">
				<a class="tcb-plain-text" href="https://www.dataquest.io/blog/topics/data-science-tutorials/">Data Science Tutorials</a>
			</div>
		</li>
			<li class="thrive-dynamic-styled-list-item tve_no_icons" data-selector=".thrive-dynamic-styled-list-item">
			<div class="tcb-styled-list-icon">
							</div>
			<div class="thrive-dynamic-styled-list-text" data-selector=".thrive-dynamic-styled-list-text a">
				<a class="tcb-plain-text" href="https://www.dataquest.io/blog/topics/dataquest-updates/">Dataquest Updates</a>
			</div>
		</li>
			<li class="thrive-dynamic-styled-list-item tve_no_icons" data-selector=".thrive-dynamic-styled-list-item">
			<div class="tcb-styled-list-icon">
							</div>
			<div class="thrive-dynamic-styled-list-text" data-selector=".thrive-dynamic-styled-list-text a">
				<a class="tcb-plain-text" href="https://www.dataquest.io/blog/topics/learning-and-motivation/">Learning and Motivation</a>
			</div>
		</li>
			<li class="thrive-dynamic-styled-list-item tve_no_icons" data-selector=".thrive-dynamic-styled-list-item">
			<div class="tcb-styled-list-icon">
							</div>
			<div class="thrive-dynamic-styled-list-text" data-selector=".thrive-dynamic-styled-list-text a">
				<a class="tcb-plain-text" href="https://www.dataquest.io/blog/topics/data-science-career-tips/">Data Science Career Tips</a>
			</div>
		</li>
			<li class="thrive-dynamic-styled-list-item tve_no_icons" data-selector=".thrive-dynamic-styled-list-item">
			<div class="tcb-styled-list-icon">
							</div>
			<div class="thrive-dynamic-styled-list-text" data-selector=".thrive-dynamic-styled-list-text a">
				<a class="tcb-plain-text" href="https://www.dataquest.io/blog/topics/student-stories/">Student Stories</a>
			</div>
		</li>
	</ul></div><div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-1714acd7792"><p data-css="tve-u-17086e6ae5d" style="">top picks</p></div><div class="tcb-post-list thrv_wrapper" data-type="list" data-pagination-type="none" data-pages_near_current="2" data-css="tve-u-16eb7cc1dec" data-total_post_count="299" data-total_sticky_count="0" data-disabled-links="1" data-no_posts_text=""><article id="post-1929" class="post-1929 post type-post status-publish format-standard has-post-thumbnail hentry category-data-science-tutorials tag-data-science tag-data-visualization tag-learn-python post-wrapper thrv_wrapper " tcb_hover_state_parent="" data-selector=".post-wrapper" data-permalink="https://www.dataquest.io/blog/how-to-communicate-with-data/"><style class="tcb-post-list-dynamic-style" type="text/css">@media (min-width: 300px){.tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-1929.post-wrapper{background-image: linear-gradient(rgba(0, 0, 0, 0.55), rgba(0, 0, 0, 0.55)), url("https://www.dataquest.io/wp-content/uploads/1-tip-data-viz.png") !important;}:not(#tve) .tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-1929.post-wrapper:hover{background-image: linear-gradient(rgba(0, 0, 0, 0.5), rgba(0, 0, 0, 0.5)), url("https://www.dataquest.io/wp-content/uploads/1-tip-data-viz.png") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-1929 [data-css="tve-u-16eb7cc1def"]{background-image: url("https://www.dataquest.io/wp-content/uploads/1-tip-data-viz.png") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-1929 [data-css="tve-u-16eb7cc1dee"]:hover [data-css="tve-u-16eb7cc1def"]{background-image: linear-gradient(rgba(0, 0, 0, 0.05), rgba(0, 0, 0, 0.05)), url("https://www.dataquest.io/wp-content/uploads/1-tip-data-viz.png") !important;}}</style>
<a href="https://www.dataquest.io/blog/how-to-communicate-with-data/" class="tve-dynamic-link" dynamic-postlink="tcb_post_the_permalink" data-shortcode-id="1929"><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-16eb7cc1dee" style="">
<div class="tve-content-box-background" data-css="tve-u-16eb7cc1def" data-tcb_hover_state_parent="1" style=""></div>
<div class="tve-cb" data-css="tve-u-16eb7cc1df0" style=""></div>
</div></a>
<div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-16eb7cd70e5"><p data-css="tve-u-16eb7cdd540"><strong>​</strong>​​​​<span class="thrive-shortcode-content" data-attr-date-format="F j, Y" data-attr-date-format-select="F j, Y" data-attr-show-time="0" data-attr-time-format="" data-attr-time-format-select="g:i a" data-attr-type="published" data-css="tve-u-17164265f0e" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_published_date" data-shortcode-name="Post Date" data-attr-link="0" data-attr-target="1" data-attr-rel="0" style="font-family: &quot;Open Sans&quot;; font-weight: 300;">February 9, 2017 </span></p></div><h2 class="tcb-post-title tcb-shortcode thrv_wrapper" data-css="tve-u-1716426374a"><a href="https://www.dataquest.io/blog/how-to-communicate-with-data/" title="1 tip for effective data visualization in Python" data-css="tve-u-1716426374a">1 tip for effective data visualization in Python</a></h2>
<div class="tve-article-cover"></div></article><article id="post-2008" class="post-2008 post type-post status-publish format-standard has-post-thumbnail hentry category-data-science-projects tag-data-science tag-data-science-projects tag-guest-post post-wrapper thrv_wrapper " tcb_hover_state_parent="" data-selector=".post-wrapper" data-permalink="https://www.dataquest.io/blog/10-data-science-projects-join/"><style class="tcb-post-list-dynamic-style" type="text/css">@media (min-width: 300px){.tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-2008.post-wrapper{background-image: linear-gradient(rgba(0, 0, 0, 0.55), rgba(0, 0, 0, 0.55)), url("https://www.dataquest.io/wp-content/uploads/2018/02/data-science-projects.jpg") !important;}:not(#tve) .tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-2008.post-wrapper:hover{background-image: linear-gradient(rgba(0, 0, 0, 0.5), rgba(0, 0, 0, 0.5)), url("https://www.dataquest.io/wp-content/uploads/2018/02/data-science-projects.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-2008 [data-css="tve-u-16eb7cc1def"]{background-image: url("https://www.dataquest.io/wp-content/uploads/2018/02/data-science-projects.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-2008 [data-css="tve-u-16eb7cc1dee"]:hover [data-css="tve-u-16eb7cc1def"]{background-image: linear-gradient(rgba(0, 0, 0, 0.05), rgba(0, 0, 0, 0.05)), url("https://www.dataquest.io/wp-content/uploads/2018/02/data-science-projects.jpg") !important;}}</style>
<a href="https://www.dataquest.io/blog/10-data-science-projects-join/" class="tve-dynamic-link" dynamic-postlink="tcb_post_the_permalink" data-shortcode-id="1929"><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-16eb7cc1dee" style="">
<div class="tve-content-box-background" data-css="tve-u-16eb7cc1def" data-tcb_hover_state_parent="1" style=""></div>
<div class="tve-cb" data-css="tve-u-16eb7cc1df0" style=""></div>
</div></a>
<div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-16eb7cd70e5"><p data-css="tve-u-16eb7cdd540"><strong>​</strong>​​​​<span class="thrive-shortcode-content" data-attr-date-format="F j, Y" data-attr-date-format-select="F j, Y" data-attr-show-time="0" data-attr-time-format="" data-attr-time-format-select="g:i a" data-attr-type="published" data-css="tve-u-17164265f0e" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_published_date" data-shortcode-name="Post Date" data-attr-link="0" data-attr-target="1" data-attr-rel="0" style="font-family: &quot;Open Sans&quot;; font-weight: 300;">February 6, 2018 </span></p></div><h2 class="tcb-post-title tcb-shortcode thrv_wrapper" data-css="tve-u-1716426374a"><a href="https://www.dataquest.io/blog/10-data-science-projects-join/" title="10 Data Science Projects You Can Join Today" data-css="tve-u-1716426374a">10 Data Science Projects You Can Join Today</a></h2>
<div class="tve-article-cover"></div></article><article id="post-8505" class="post-8505 post type-post status-publish format-standard has-post-thumbnail hentry category-data-science-career-tips tag-careers post-wrapper thrv_wrapper " tcb_hover_state_parent="" data-selector=".post-wrapper" data-permalink="https://www.dataquest.io/blog/10-data-analytics-jobs/"><style class="tcb-post-list-dynamic-style" type="text/css">@media (min-width: 300px){.tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-8505.post-wrapper{background-image: linear-gradient(rgba(0, 0, 0, 0.55), rgba(0, 0, 0, 0.55)), url("https://www.dataquest.io/wp-content/uploads/2019/02/data-analytics.jpg") !important;}:not(#tve) .tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-8505.post-wrapper:hover{background-image: linear-gradient(rgba(0, 0, 0, 0.5), rgba(0, 0, 0, 0.5)), url("https://www.dataquest.io/wp-content/uploads/2019/02/data-analytics.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-8505 [data-css="tve-u-16eb7cc1def"]{background-image: url("https://www.dataquest.io/wp-content/uploads/2019/02/data-analytics.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-8505 [data-css="tve-u-16eb7cc1dee"]:hover [data-css="tve-u-16eb7cc1def"]{background-image: linear-gradient(rgba(0, 0, 0, 0.05), rgba(0, 0, 0, 0.05)), url("https://www.dataquest.io/wp-content/uploads/2019/02/data-analytics.jpg") !important;}}</style>
<a href="https://www.dataquest.io/blog/10-data-analytics-jobs/" class="tve-dynamic-link" dynamic-postlink="tcb_post_the_permalink" data-shortcode-id="1929"><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-16eb7cc1dee" style="">
<div class="tve-content-box-background" data-css="tve-u-16eb7cc1def" data-tcb_hover_state_parent="1" style=""></div>
<div class="tve-cb" data-css="tve-u-16eb7cc1df0" style=""></div>
</div></a>
<div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-16eb7cd70e5"><p data-css="tve-u-16eb7cdd540"><strong>​</strong>​​​​<span class="thrive-shortcode-content" data-attr-date-format="F j, Y" data-attr-date-format-select="F j, Y" data-attr-show-time="0" data-attr-time-format="" data-attr-time-format-select="g:i a" data-attr-type="published" data-css="tve-u-17164265f0e" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_published_date" data-shortcode-name="Post Date" data-attr-link="0" data-attr-target="1" data-attr-rel="0" style="font-family: &quot;Open Sans&quot;; font-weight: 300;">February 12, 2019 </span></p></div><h2 class="tcb-post-title tcb-shortcode thrv_wrapper" data-css="tve-u-1716426374a"><a href="https://www.dataquest.io/blog/10-data-analytics-jobs/" title="10 High-Paying Jobs That Require a Knowledge of Data Analytics" data-css="tve-u-1716426374a">10 High-Paying Jobs That Require a Knowledge of Data Analytics</a></h2>
<div class="tve-article-cover"></div></article><article id="post-2062" class="post-2062 post type-post status-publish format-standard has-post-thumbnail hentry category-learning-and-motivation tag-data-science-projects tag-resources tag-tutorials post-wrapper thrv_wrapper " tcb_hover_state_parent="" data-selector=".post-wrapper" data-permalink="https://www.dataquest.io/blog/design-tips-for-data-viz/"><style class="tcb-post-list-dynamic-style" type="text/css">@media (min-width: 300px){.tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-2062.post-wrapper{background-image: linear-gradient(rgba(0, 0, 0, 0.55), rgba(0, 0, 0, 0.55)), url("https://www.dataquest.io/wp-content/uploads/2018/10/design.jpg") !important;}:not(#tve) .tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-2062.post-wrapper:hover{background-image: linear-gradient(rgba(0, 0, 0, 0.5), rgba(0, 0, 0, 0.5)), url("https://www.dataquest.io/wp-content/uploads/2018/10/design.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-2062 [data-css="tve-u-16eb7cc1def"]{background-image: url("https://www.dataquest.io/wp-content/uploads/2018/10/design.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-2062 [data-css="tve-u-16eb7cc1dee"]:hover [data-css="tve-u-16eb7cc1def"]{background-image: linear-gradient(rgba(0, 0, 0, 0.05), rgba(0, 0, 0, 0.05)), url("https://www.dataquest.io/wp-content/uploads/2018/10/design.jpg") !important;}}</style>
<a href="https://www.dataquest.io/blog/design-tips-for-data-viz/" class="tve-dynamic-link" dynamic-postlink="tcb_post_the_permalink" data-shortcode-id="1929"><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-16eb7cc1dee" style="">
<div class="tve-content-box-background" data-css="tve-u-16eb7cc1def" data-tcb_hover_state_parent="1" style=""></div>
<div class="tve-cb" data-css="tve-u-16eb7cc1df0" style=""></div>
</div></a>
<div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-16eb7cd70e5"><p data-css="tve-u-16eb7cdd540"><strong>​</strong>​​​​<span class="thrive-shortcode-content" data-attr-date-format="F j, Y" data-attr-date-format-select="F j, Y" data-attr-show-time="0" data-attr-time-format="" data-attr-time-format-select="g:i a" data-attr-type="published" data-css="tve-u-17164265f0e" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_published_date" data-shortcode-name="Post Date" data-attr-link="0" data-attr-target="1" data-attr-rel="0" style="font-family: &quot;Open Sans&quot;; font-weight: 300;">October 25, 2018 </span></p></div><h2 class="tcb-post-title tcb-shortcode thrv_wrapper" data-css="tve-u-1716426374a"><a href="https://www.dataquest.io/blog/design-tips-for-data-viz/" title="11 Design Tips for Data Visualization" data-css="tve-u-1716426374a">11 Design Tips for Data Visualization</a></h2>
<div class="tve-article-cover"></div></article></div>

</div></aside><div class="main-columns-separator"></div>
<div id="theme-content-section" class="theme-section content-section"><div class="section-background"></div><div class="section-content">
<div class="tcb-clear" data-css="tve-u-17019a652a1"><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-17019a634b0" data-tve-scroll="{&quot;disabled&quot;:[],&quot;top&quot;:12,&quot;mode&quot;:&quot;sticky&quot;,&quot;end&quot;:&quot;element&quot;,&quot;el_id&quot;:&quot;Divider&quot;}">
	<div class="tve-content-box-background" style="" data-css="tve-u-17063360500"></div>
	<div class="tve-cb"><div class="tcb-clear" data-css="tve-u-17019ad95f7"><div class="thrv_wrapper thrv_icon tcb-icon-display" data-css="tve-u-170199f5045" style="" data-float-m="1"><svg class="tcb-icon" viewBox="0 0 448 512" data-id="icon-share-alt-solid" data-name="" style="">
            <path d="M352 320c-22.608 0-43.387 7.819-59.79 20.895l-102.486-64.054a96.551 96.551 0 0 0 0-41.683l102.486-64.054C308.613 184.181 329.392 192 352 192c53.019 0 96-42.981 96-96S405.019 0 352 0s-96 42.981-96 96c0 7.158.79 14.13 2.276 20.841L155.79 180.895C139.387 167.819 118.608 160 96 160c-53.019 0-96 42.981-96 96s42.981 96 96 96c22.608 0 43.387-7.819 59.79-20.895l102.486 64.054A96.301 96.301 0 0 0 256 416c0 53.019 42.981 96 96 96s96-42.981 96-96-42.981-96-96-96z"></path>
        </svg></div></div><div class="thrv_wrapper thrv_social_custom thrv_social tcb-mobile-hidden" data-counts="" data-min_shares="0" data-css="tve-u-170199bd58e" style="" data-class=" tcb-mobile-hidden tcb-mobile-hidden">
<div class="tve_social_items tve_social_custom tve_style_5 tve_social_ib" style="font-size: 8px !important;">
<div class="tve_s_item tve_s_fb_share" data-s="fb_share" data-href="https://www.dataquest.io/blog/jupyter-notebook-tutorial/" data-label="Share">
	<a href="javascript:void(0)" class="tve_s_link"><span class="tve_s_icon"></span><span class="tve_s_text">Share</span><span class="tve_s_count">0</span></a>
</div><div class="tve_s_item tve_s_t_share" data-s="t_share" data-href="https://www.dataquest.io/blog/jupyter-notebook-tutorial/" data-label="Tweet">
	<a href="javascript:void(0)" class="tve_s_link"><span class="tve_s_icon"></span><span class="tve_s_text">Tweet</span><span class="tve_s_count">0</span></a>
</div><div class="tve_s_item tve_s_in_share" data-s="in_share" data-href="https://www.dataquest.io/blog/jupyter-notebook-tutorial/" data-label="Share">
	<a href="javascript:void(0)" class="tve_s_link"><span class="tve_s_icon"></span><span class="tve_s_text">Share</span><span class="tve_s_count">0</span></a>
</div>
</div>
</div><div class="thrv_wrapper thrv_social_custom thrv_social tcb-desktop-hidden tcb-tablet-hidden" data-counts="" data-min_shares="0" data-css="tve-u-17019ad2ca6" style="" data-class=" tcb-desktop-hidden tcb-tablet-hidden tcb-desktop-hidden tcb-tablet-hidden">
<div class="tve_social_items tve_social_custom tve_style_5 tve_social_ib" style="">
<div class="tve_s_item tve_s_fb_share" data-s="fb_share" data-href="https://www.dataquest.io/blog/jupyter-notebook-tutorial/" data-label="Share">
	<a href="javascript:void(0)" class="tve_s_link"><span class="tve_s_icon"></span><span class="tve_s_text">Share</span><span class="tve_s_count">0</span></a>
</div><div class="tve_s_item tve_s_t_share" data-s="t_share" data-href="https://www.dataquest.io/blog/jupyter-notebook-tutorial/" data-label="Tweet">
	<a href="javascript:void(0)" class="tve_s_link"><span class="tve_s_icon"></span><span class="tve_s_text">Tweet</span><span class="tve_s_count">0</span></a>
</div><div class="tve_s_item tve_s_in_share" data-s="in_share" data-href="https://www.dataquest.io/blog/jupyter-notebook-tutorial/" data-label="Share">
	<a href="javascript:void(0)" class="tve_s_link"><span class="tve_s_icon"></span><span class="tve_s_text">Share</span><span class="tve_s_count">0</span></a>
</div>
</div>
</div></div>
</div></div><section class="tcb-post-content tcb-shortcode thrv_wrapper" data-css="tve-u-16ec5d248bb"><div class="kg-card-markdown">
<p>
    The Jupyter Notebook is an incredibly powerful tool for interactively developing and presenting data science projects. This article will walk you through how to set up Jupyter Notebooks on your local machine and how to start using it to do data science projects.
  </p>
<p>
    First, though: what is a &#8220;notebook&#8221;? A notebook integrates code and its output into a single document that combines visualizations, narrative text, mathematical equations, and other rich media. This intuitive workflow promotes iterative and rapid development, making notebooks an increasingly popular choice at the heart of contemporary data science, analysis, and increasingly science at large.
  </p>
<p>
    Best of all, as part of the open source <a href="https://jupyter.org/" draggable="false">Project Jupyter</a>, they are completely free.
  </p>
<p>
    The Jupyter project is the successor to the earlier IPython Notebook, which was first published as a prototype in 2010. Although it is possible to use many different programming languages within Jupyter Notebooks, this article will focus on Python as it is the most common use case. (Among R users, <a href="https://www.rstudio.com/">R Studio</a> tends to be a more popular choice).
  </p>
<p>
    To get the most out of this tutorial you should be familiar with programming, specifically Python and <a href="https://pandas.pydata.org/" draggable="false">pandas</a> specifically. That said, if you have experience with another language, the Python in this article shouldn&#8217;t be too cryptic, and will still help you get Jupyter Notebooks set up locally. Jupyter Notebooks can also act as a flexible platform for getting to grips with pandas and even Python, as will become apparent in this article.
  </p>
<p>
    We will:
  </p>
<ul>
<li>
      Cover the basics of installing Jupyter and creating your first notebook
    </li>
<li>
      Delve deeper and learn all the important terminology
    </li>
<li>
      Explore how easily notebooks can be shared and published online. Indeed, this article <em>is</em> a Jupyter Notebook! Everything here was written in the Jupyter Notebook environment, though you are viewing it in a read-only form.
    </li>
</ul>
<h3 id="exampledataanalysisinajupyternotebook">
    Example Data Analysis in a Jupyter Notebook<br />
  </h3>
<p>
    First, we will walk through setup and a sample analysis to answer a real-life question. This will demonstrate how the flow of a notebook makes data science tasks more intuitive for us as we work, and for others once it&#8217;s time to share our work.
  </p>
<p>
    So, let&#8217;s say you&#8217;re a data analyst and you&#8217;ve been tasked with finding out how the profits of the largest companies in the US changed historically. You find a data set of Fortune 500 companies spanning over 50 years since the list&#8217;s first publication in 1955, put together from <a href="http://archive.fortune.com/magazines/fortune/fortune500_archive/full/2005/" draggable="false">Fortune&#8217;s public archive</a>. We&#8217;ve gone ahead and created a CSV of the data you can use <a href="https://s3.amazonaws.com/dq-blog-files/fortune500.csv" draggable="false">here</a>.
  </p>
<p>
    As we shall demonstrate, Jupyter Notebooks are perfectly suited for this investigation. First, let&#8217;s go ahead and install Jupyter.
  </p>
</div>
<h2 id="installation">Installation</h2>
<p>The easiest way for a beginner to get started with Jupyter Notebooks is by installing <a href="https://www.anaconda.com/distribution/">Anaconda</a>. Anaconda is the most widely used Python distribution for data science and comes pre-loaded with all the most popular libraries and tools. Some of the biggest Python libraries wrapped up in Anaconda include <a href="http://www.numpy.org/">NumPy</a>, <a href="https://pandas.pydata.org/">pandas</a> and <a href="https://matplotlib.org/">Matplotlib</a>, though the <a href="https://docs.anaconda.com/anaconda/packages/pkg-docs">full 1000+ list</a> is exhaustive. This lets you hit the ground running in your own fully stocked data science workshop without the hassle of managing countless installations or worrying about dependencies and OS-specific (read: Windows-specific) installation issues.</p>
<p>To get Anaconda, simply:</p>
<ol>
<li><a href="https://www.anaconda.com/download/">Download</a> the latest version of Anaconda for Python 3 (ignore Python 2.7).</li>
<li>Install Anaconda by following the instructions on the download page and/or in the executable.</li>
</ol>
<p>If you are a more advanced user with Python already installed and prefer to manage your packages manually, you can just use pip:</p>
<pre class="language-bash"><code class="language-bash">pip3 install jupyter</code></pre>
<h2 id="creatingyourfirstnotebook">Creating Your First Notebook</h2>
<p>In this section, we&#8217;re going to learn to run and save notebooks, familiarize ourselves with their structure, and understand the interface. We&#8217;ll become intimate with some core terminology that will steer you towards a practical understanding of how to use Jupyter Notebooks by yourself and set us up for the next section, which steps through an example data analysis and brings everything we learn here to life.</p>
<h3 id="runningjupyter">Running Jupyter</h3>
<p>On Windows, you can run Jupyter via the shortcut Anaconda adds to your start menu, which will open a new tab in your default web browser that should look something like the following screenshot.</p>
<p><noscript><img src="/wp-content/uploads/2019/01/jupyter-dashboard.jpg" alt="Jupyter control panel" /></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="/wp-content/uploads/2019/01/jupyter-dashboard.jpg" alt="Jupyter control panel" /></p>
<p>This isn&#8217;t a notebook just yet, but don&#8217;t panic! There&#8217;s not much to it. This is the Notebook Dashboard, specifically designed for managing your Jupyter Notebooks. Think of it as the launchpad for exploring, editing and creating your notebooks.</p>
<p>Be aware that the dashboard will give you access only to the files and sub-folders contained within Jupyter&#8217;s start-up directory; however, the start-up directory <a href="https://stackoverflow.com/q/35254852/604687">can be changed</a>. It is also possible to start the dashboard on any system via the command prompt (or terminal on Unix systems) by entering the command <code>jupyter notebook</code>; in this case, the current working directory will be the start-up directory.</p>
<p>The astute reader may have noticed that the URL for the dashboard is something like <code>http://localhost:8888/tree</code>. Localhost is not a website, but indicates that the content is being served from your <em>local</em> machine: your own computer. Jupyter&#8217;s Notebooks and dashboard are web apps, and Jupyter starts up a local Python server to serve these apps to your web browser, making it essentially platform independent and opening the door to easier sharing on the web.</p>
<p>The dashboard&#8217;s interface is mostly self-explanatory — though we will come back to it briefly later. So what are we waiting for? Browse to the folder in which you would like to create your first notebook, click the &#8220;New&#8221; drop-down button in the top-right and select &#8220;Python 3&#8221; (or the version of your choice).</p>
<p><noscript><img src="/wp-content/uploads/2019/01/new-notebook-menu.jpg" alt="New notebook menu" /></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="/wp-content/uploads/2019/01/new-notebook-menu.jpg" alt="New notebook menu" /></p>
<p>Hey presto, here we are! Your first Jupyter Notebook will open in new tab — each notebook uses its own tab because you can open multiple notebooks simultaneously. If you switch back to the dashboard, you will see the new file <code>Untitled.ipynb</code> and you should see some green text that tells you your notebook is running.</p>
<h4 id="whatisanipynbfile">What is an ipynb File?</h4>
<p>It will be useful to understand what this file really is. Each <code>.ipynb</code> file is a text file that describes the contents of your notebook in a format called <a href="https://en.wikipedia.org/wiki/JSON">JSON</a>. Each cell and its contents, including image attachments that have been converted into strings of text, is listed therein along with some <a href="https://ipython.org/ipython-doc/3/notebook/nbformat.html#metadata">metadata</a>. You can edit this yourself — if you know what you are doing! — by selecting &#8220;Edit > Edit Notebook Metadata&#8221; from the menu bar in the notebook.</p>
<p>You can also view the contents of your notebook files by selecting &#8220;Edit&#8221; from the controls on the dashboard, but the keyword here is &#8220;<em>can</em>&#8220;; there&#8217;s no reason other than curiosity to do so unless you really know what you are doing.</p>
<h3 id="thenotebookinterface">The Notebook Interface</h3>
<p>Now that you have an open notebook in front of you, its interface will hopefully not look entirely alien; after all, Jupyter is essentially just an advanced word processor. Why not take a look around? Check out the menus to get a feel for it, especially take a few moments to scroll down the list of commands in the command palette, which is the small button with the keyboard icon (or <code>Ctrl + Shift + P</code>).</p>
<p><noscript><img src="/wp-content/uploads/2019/01/new-notebook.jpg" alt="New Jupyter Notebook" /></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="/wp-content/uploads/2019/01/new-notebook.jpg" alt="New Jupyter Notebook" /></p>
<p>There are two fairly prominent terms that you should notice, which are probably new to you: <em>cells</em> and <em>kernels</em> are key both to understanding Jupyter and to what makes it more than just a word processor. Fortunately, these concepts are not difficult to understand.</p>
<ul>
<li>A kernel is a &#8220;computational engine&#8221; that executes the code contained in a notebook document.</li>
<li>A cell is a container for text to be displayed in the notebook or code to be executed by the notebook&#8217;s kernel.</li>
</ul>
<h3 id="cells">Cells</h3>
<p>We&#8217;ll return to kernels a little later, but first let&#8217;s come to grips with cells. Cells form the body of a notebook. In the screenshot of a new notebook in the section above, that box with the green outline is an empty cell. There are two main cell types that we will cover:</p>
<ul>
<li>A <strong>code cell</strong> contains code to be executed in the kernel and displays its output below.</li>
<li>A <strong>Markdown cell</strong> contains text formatted using Markdown and displays its output in-place when it is run.</li>
</ul>
<p>The first cell in a new notebook is always a code cell. Let&#8217;s test it out with a classic hello world example. Type <code>print('Hello World!')</code> into the cell and click the run button <noscript><img src="/wp-content/uploads/2019/01/run-button.jpg" alt="Notebook Run Button" style="display: inline-block; margin: 0; height: 1.5em;" /></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="/wp-content/uploads/2019/01/run-button.jpg" alt="Notebook Run Button" style="display: inline-block; margin: 0; height: 1.5em;" /> in the toolbar above or press <code>Ctrl + Enter</code>. The result should look like this:</p>
<pre class="language-bash"><code class="language-bash">print('Hello World!')</code></pre>
<pre class="language-bash"><code class="language-bash">Hello World!</code></pre>
<p>When you ran the cell, its output will have been displayed below and the label to its left will have changed from <code>In [ ]</code> to <code>In [1]</code>. The output of a code cell also forms part of the document, which is why you can see it in this article. You can always tell the difference between code and Markdown cells because code cells have that label on the left and Markdown cells do not.</p>
<p>The &#8220;In&#8221; part of the label is simply short for &#8220;Input,&#8221; while the label number indicates when the cell was executed on the kernel — in this case the cell was executed first. Run the cell again and the label will change to <code>In [2]</code> because now the cell was the second to be run on the kernel. It will become clearer why this is so useful later on when we take a closer look at kernels.</p>
<p>From the menu bar, click <em>Insert</em> and select <em>Insert Cell Below</em> to create a new code cell underneath your first and try out the following code to see what happens. Do you notice anything different?</p>
<pre class="language-bash"><code class="language-bash">
import time
time.sleep(3)
</code></pre>
<p>This cell doesn&#8217;t produce any output, but it does take three seconds to execute. Notice how Jupyter signifies that the cell is currently running by changing its label to <code>In [*]</code>.</p>
<p>In general, the output of a cell comes from any text data specifically printed during the cells execution, as well as the value of the last line in the cell, be it a lone variable, a function call, or something else. For example:</p>
<pre><code class="language-python">
def say_hello(recipient):
    return 'Hello, {}!'.format(recipient)
say_hello('Tim')
</code></pre>
<pre class="language-bash"><code class="language-bash">
'Hello, Tim!'
</code></pre>
<p>You&#8217;ll find yourself using this almost constantly in your own projects, and we&#8217;ll see more of it later on.</p>
<h3 id="keyboardshortcuts">Keyboard Shortcuts</h3>
<p>One final thing you may have observed when running your cells is that their border turned blue, whereas it was green while you were editing. There is always one &#8220;active&#8221; cell highlighted with a border whose color denotes its current mode, where green means &#8220;edit mode&#8221; and blue is &#8220;command mode.&#8221;</p>
<p>So far we have seen how to run a cell with <code>Ctrl + Enter</code>, but there are plenty more. Keyboard shortcuts are a very popular aspect of the Jupyter environment because they facilitate a speedy cell-based workflow. Many of these are actions you can carry out on the active cell when it&#8217;s in command mode.</p>
<p>Below, you&#8217;ll find a list of some of Jupyter&#8217;s keyboard shortcuts. You&#8217;re not expected to pick them up immediately, but the list should give you a good idea of what&#8217;s possible.</p>
<ul>
<li>Toggle between edit and command mode with <code>Esc</code> and <code>Enter</code>, respectively.</li>
<li>Once in command mode:
<ul>
<li>Scroll up and down your cells with your <code>Up</code> and <code>Down</code> keys.</li>
<li>Press <code>A</code> or <code>B</code> to insert a new cell above or below the active cell.</li>
<li><code>M</code> will transform the active cell to a Markdown cell.</li>
<li><code>Y</code> will set the active cell to a code cell.</li>
<li><code>D + D</code> (<code>D</code> twice) will delete the active cell.</li>
<li><code>Z</code> will undo cell deletion.</li>
<li>Hold <code>Shift</code> and press <code>Up</code> or <code>Down</code> to select multiple cells at once.
<ul>
<li>With multiple cells selected, <code>Shift + M</code> will merge your selection.</li>
</ul>
</li>
</ul>
</li>
<li><code>Ctrl + Shift + -</code>, in edit mode, will split the active cell at the cursor.</li>
<li>You can also click and <code>Shift + Click</code> in the margin to the left of your cells to select them.</li>
</ul>
<p>Go ahead and try these out in your own notebook. Once you&#8217;ve had a play, create a new Markdown cell and we&#8217;ll learn how to format the text in our notebooks.</p>
<h3 id="markdown">Markdown</h3>
<p><a href="https://www.markdownguide.org/">Markdown</a> is a lightweight, easy to learn markup language for formatting plain text. Its syntax has a one-to-one correspondance with HTML tags, so some prior knowledge here would be helpful but is definitely not a prerequisite. Remember that this article was written in a Jupyter notebook, so all of the narrative text and images you have seen so far was achieved in Markdown. Let&#8217;s cover the basics with a quick example.</p>
<pre class="language-bash"><code class="language-bash">
# This is a level 1 heading
## This is a level 2 heading
This is some plain text that forms a paragraph.
Add emphasis via **bold** and __bold__, or *italic* and _italic_.
Paragraphs must be separated by an empty line.
* Sometimes we want to include lists.
* Which can be indented.
1. Lists can also be numbered.
2. For ordered lists.
[It is possible to include hyperlinks](https://www.example.com)
Inline code uses single backticks: `foo()`, and code blocks use triple backticks:
```
bar()
```
Or can be indented by 4 spaces:
    foo()
And finally, adding images is easy: ![Alt text](https://www.example.com/image.jpg)
</code></pre>
<p>When attaching images, you have three options:</p>
<ul>
<li>Use a URL to an image on the web.</li>
<li>Use a local URL to an image that you will be keeping alongside your notebook, such as in the same git repo.</li>
<li>
<p>Add an attachment via &#8220;Edit > Insert Image&#8221;; this will convert the image into a string and store it inside your notebook <code>.ipynb</code> file.</p>
</li>
<li>
<p>Note that this will make your <code>.ipynb</code> file much larger!</p>
</li>
</ul>
<p>There is plenty more detail to Markdown, especially around hyperlinking, and it&#8217;s also possible to simply include plain HTML. Once you find yourself pushing the limits of the basics above, you can refer to the <a href="https://daringfireball.net/projects/markdown/syntax">official guide</a> from the creator, John Gruber, on his website.</p>
<h3 id="kernels">Kernels</h3>
<p>Behind every notebook runs a kernel. When you run a code cell, that code is executed within the kernel and any output is returned back to the cell to be displayed. The kernel&#8217;s state persists over time and between cells — it pertains to the document as a whole and not individual cells.</p>
<p>For example, if you import libraries or declare variables in one cell, they will be available in another. In this way, you can think of a notebook document as being somewhat comparable to a script file, except that it is multimedia. Let&#8217;s try this out to get a feel for it. First, we&#8217;ll import a Python package and define a function.</p>
<pre><code class="language-python">
import numpy as np
def square(x):
    return x * x
</code></pre>
<p>Once we&#8217;ve executed the cell above, we can reference <code>np</code> and <code>square</code> in any other cell.</p>
<pre><code class="language-python">
x = np.random.randint(1, 10)
y = square(x)
print('%d squared is %d' % (x, y))</code></pre>
<pre class="language-bash"><code class="language-bash">
1 squared is 1
</code></pre>
<p>This will work regardless of the order of the cells in your notebook. You can try it yourself, let&#8217;s print out our variables again.</p>
<pre class="language-bash"><code class="language-bash">
print('Is %d squared is %d?' % (x, y))
</code></pre>
<pre class="language-bash"><code class="language-bash">
Is 1 squared is 1?
</code></pre>
<p>No surprises here! But now let&#8217;s change <code>y</code>.</p>
<pre class="language-bash"><code class="language-bash">
y = 10
</code></pre>
<p>What do you think will happen if we run the cell containing our <code>print</code> statement again? We will get the output <code>Is 4 squared is 10?</code>!</p>
<p>Most of the time, the flow in your notebook will be top-to-bottom, but it&#8217;s common to go back to make changes. In this case, the order of execution stated to the left of each cell, such as <code>In [6]</code>, will let you know whether any of your cells have stale output. And if you ever wish to reset things, there are several incredibly useful options from the Kernel menu:</p>
<ul>
<li>Restart: restarts the kernel, thus clearing all the variables etc that were defined.</li>
<li>Restart &amp; Clear Output: same as above but will also wipe the output displayed below your code cells.</li>
<li>Restart &amp; Run All: same as above but will also run all your cells in order from first to last.</li>
</ul>
<p>If your kernel is ever stuck on a computation and you wish to stop it, you can choose the Interupt option.</p>
<h4 id="choosingakernel">Choosing a Kernel</h4>
<p>You may have noticed that Jupyter gives you the option to change kernel, and in fact there are many different options to choose from. Back when you created a new notebook from the dashboard by selecting a Python version, you were actually choosing which kernel to use.</p>
<p>Not only are there kernels for different versions of Python, but also for <a href="https://github.com/jupyter/jupyter/wiki/Jupyter-kernels">over 100 languages</a> including Java, C, and even Fortran. Data scientists may be particularly interested in the kernels for <a href="https://irkernel.github.io/">R</a> and <a href="https://github.com/JuliaLang/IJulia.jl">Julia</a>, as well as both <a href="https://github.com/imatlab/imatlab">imatlab</a> and the <a href="https://github.com/calysto/matlab_kernel">Calysto MATLAB Kernel</a> for Matlab. The <a href="https://github.com/vatlab/SOS">SoS kernel</a> provides multi-language support within a single notebook. Each kernel has its own installation instructions, but will likely require you to run some commands on your computer.</p>
<h2 id="exampleanalysis">Example Analysis</h2>
<p>Now we&#8217;ve looked at <em>what</em> a Jupyter Notebook is, it&#8217;s time to look at <em>how</em> they&#8217;re used in practice, which should give you a clearer understanding of <em>why</em> they are so popular. It&#8217;s finally time to get started with that Fortune 500 data set mentioned earlier. Remember, our goal is to find out how the profits of the largest companies in the US changed historically.</p>
<p>It&#8217;s worth noting that everyone will develop their own preferences and style, but the general principles still apply, and you can follow along with this section in your own notebook if you wish, which gives you the scope to play around.</p>
<h3 id="namingyournotebooks">Naming Your Notebooks</h3>
<p>Before you start writing your project, you&#8217;ll probably want to give it a meaningful name. Perhaps somewhat confusingly, you cannot name or rename your notebooks from the notebook app itself, but must use either the dashboard or your file browser to rename the <code>.ipynb</code> file. We&#8217;ll head back to the dashboard to rename the file you created earlier, which will have the default notebook file name <code>Untitled.ipynb</code>.</p>
<p>You cannot rename a notebook while it is running, so you&#8217;ve first got to shut it down. The easiest way to do this is to select &#8220;File > Close and Halt&#8221; from the notebook menu. However, you can also shutdown the kernel either by going to &#8220;Kernel > Shutdown&#8221; from within the notebook app or by selecting the notebook in the dashboard and clicking &#8220;Shutdown&#8221; (see image below).</p>
<p><noscript><img src="/wp-content/uploads/2019/01/notebook-running.jpg" alt="A running notebook" /></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="/wp-content/uploads/2019/01/notebook-running.jpg" alt="A running notebook" /></p>
<p>You can then select your notebook and and click &#8220;Rename&#8221; in the dashboard controls.</p>
<p><noscript><img src="/wp-content/uploads/2019/01/notebook-controls.jpg" alt="A running notebook" /></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="/wp-content/uploads/2019/01/notebook-controls.jpg" alt="A running notebook" /></p>
<p>Note that closing the notebook tab in your browser will <strong>not</strong> &#8220;close&#8221; your notebook in the way closing a document in a traditional application will. The notebook&#8217;s kernel will continue to run in the background and needs to be shut down before it is truly &#8220;closed&#8221; — though this is pretty handy if you accidentally close your tab or browser! If the kernel is shut down, you can close the tab without worrying about whether it is still running or not.</p>
<p>Once you&#8217;ve named your notebook, open it back up and we&#8217;ll get going.</p>
<h3 id="setup">Setup</h3>
<p>It&#8217;s common to start off with a code cell specifically for imports and setup, so that if you choose to add or change anything, you can simply edit and re-run the cell without causing any side-effects.</p>
<pre><code class="language-python">

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
</code></pre>
<p>We import <a href="https://pandas.pydata.org/">pandas</a> to work with our data, <a href="https://matplotlib.org/">Matplotlib</a> to plot charts, and <a href="https://seaborn.pydata.org/">Seaborn</a> to make our charts prettier. It&#8217;s also common to import <a href="http://www.numpy.org/">NumPy</a> but in this case, although we use it via pandas, we don&#8217;t need to explicitly. And that first line isn&#8217;t a Python command, but uses something called a line magic to instruct Jupyter to capture Matplotlib plots and render them in the cell output; this is one of a range of advanced features that are out of the scope of this article.</p>
<p>Let&#8217;s go ahead and load our data.</p>
<pre class="language-python"><code class="language-python">
df = pd.read_csv('fortune500.csv')
</code></pre>
<p>It&#8217;s sensible to also do this in a single cell in case we need to reload it at any point.</p>
<h3 id="saveandcheckpoint">Save and Checkpoint</h3>
<p>Now we&#8217;ve got started, it&#8217;s best practice to save regularly. Pressing <code>Ctrl + S</code> will save your notebook by calling the &#8220;Save and Checkpoint&#8221; command, but what this checkpoint thing?</p>
<p>Every time you create a new notebook, a checkpoint file is created as well as your notebook file; it will be located within a hidden subdirectory of your save location called <code>.ipynb_checkpoints</code> and is also a <code>.ipynb</code> file. By default, Jupyter will autosave your notebook every 120 seconds to this checkpoint file without altering your primary notebook file. When you &#8220;Save and Checkpoint,&#8221; both the notebook and checkpoint files are updated. Hence, the checkpoint enables you to recover your unsaved work in the event of an unexpected issue. You can revert to the checkpoint from the menu via &#8220;File > Revert to Checkpoint.&#8221;<br />
<span style="display:none" class="tl-placeholder-f-type-shortcode_16321"></span></p>
<h3 id="investigatingourdataset">Investigating Our Data Set</h3>
<p>Now we&#8217;re really rolling! Our notebook is safely saved and we&#8217;ve loaded our data set <code>df</code> into the most-used pandas data structure, which is called a <code>DataFrame</code> and basically looks like a table. What does ours look like?</p>
<pre class="language-python"><code class="language-python">df.head()</code></pre>
<div style="max-width: 100%;">
<style scoped="">
    .dataframe tbody tr th:only-of-type { vertical-align: middle; }</p>
<pre><code>.dataframe tbody tr th { vertical-align: top;}.dataframe thead th { text-align: right;}</code></pre>
</style>
<table border="1" class="dataframe">
<thead>
<tr style="text-align: right;">
<th>
        </th>
<th>
          Year
        </th>
<th>
          Rank
        </th>
<th>
          Company
        </th>
<th>
          Revenue (in millions)
        </th>
<th>
          Profit (in millions)
        </th>
</tr>
</thead>
<tbody>
<tr>
<th>
          0
        </th>
<td>
          1955
        </td>
<td>
          1
        </td>
<td>
          General Motors
        </td>
<td>
          9823.5
        </td>
<td>
          806
        </td>
</tr>
<tr>
<th>
          1
        </th>
<td>
          1955
        </td>
<td>
          2
        </td>
<td>
          Exxon Mobil
        </td>
<td>
          5661.4
        </td>
<td>
          584.8
        </td>
</tr>
<tr>
<th>
          2
        </th>
<td>
          1955
        </td>
<td>
          3
        </td>
<td>
          U.S. Steel
        </td>
<td>
          3250.4
        </td>
<td>
          195.4
        </td>
</tr>
<tr>
<th>
          3
        </th>
<td>
          1955
        </td>
<td>
          4
        </td>
<td>
          General Electric
        </td>
<td>
          2959.1
        </td>
<td>
          212.6
        </td>
</tr>
<tr>
<th>
          4
        </th>
<td>
          1955
        </td>
<td>
          5
        </td>
<td>
          Esmark
        </td>
<td>
          2510.8
        </td>
<td>
          19.1
        </td>
</tr>
</tbody>
</table>
</div>
<pre class="language-python"><code class="language-python">df.tail()</code></pre>
<div style="max-width: 100%;">
<style scoped="">
    .dataframe tbody tr th:only-of-type { vertical-align: middle; }</p>
<pre><code>.dataframe tbody tr th { vertical-align: top;}.dataframe thead th { text-align: right;}</code></pre>
</style>
<table border="1" class="dataframe">
<thead>
<tr style="text-align: right;">
<th>
        </th>
<th>
          Year
        </th>
<th>
          Rank
        </th>
<th>
          Company
        </th>
<th>
          Revenue (in millions)
        </th>
<th>
          Profit (in millions)
        </th>
</tr>
</thead>
<tbody>
<tr>
<th>
          25495
        </th>
<td>
          2005
        </td>
<td>
          496
        </td>
<td>
          Wm. Wrigley Jr.
        </td>
<td>
          3648.6
        </td>
<td>
          493
        </td>
</tr>
<tr>
<th>
          25496
        </th>
<td>
          2005
        </td>
<td>
          497
        </td>
<td>
          Peabody Energy
        </td>
<td>
          3631.6
        </td>
<td>
          175.4
        </td>
</tr>
<tr>
<th>
          25497
        </th>
<td>
          2005
        </td>
<td>
          498
        </td>
<td>
          Wendy&#8217;s International
        </td>
<td>
          3630.4
        </td>
<td>
          57.8
        </td>
</tr>
<tr>
<th>
          25498
        </th>
<td>
          2005
        </td>
<td>
          499
        </td>
<td>
          Kindred Healthcare
        </td>
<td>
          3616.6
        </td>
<td>
          70.6
        </td>
</tr>
<tr>
<th>
          25499
        </th>
<td>
          2005
        </td>
<td>
          500
        </td>
<td>
          Cincinnati Financial
        </td>
<td>
          3614.0
        </td>
<td>
          584
        </td>
</tr>
</tbody>
</table>
</div>
<p>Looking good. We have the columns we need, and each row corresponds to a single company in a single year.</p>
<p>Let&#8217;s just rename those columns so we can refer to them later.</p>
<pre class="language-python"><code class="language-python">df.columns = ['year', 'rank', 'company', 'revenue', 'profit']</code></pre>
<p>Next, we need to explore our data set. Is it complete? Did pandas read it as expected? Are any values missing?</p>
<pre class="language-python"><code class="language-python">len(df)</code></pre>
<pre class="language-python"><code class="language-python">25500</code></pre>
<p>Okay, that looks good — that&#8217;s 500 rows for every year from 1955 to 2005, inclusive.</p>
<p>Let&#8217;s check whether our data set has been imported as we would expect. A simple check is to see if the data types (or dtypes) have been correctly interpreted.</p>
<pre class="language-python"><code class="language-python">df.dtypes</code></pre>
<pre class="language-python"><code class="language-python">
year int64
rank int64
company object
revenue float64
profit object
dtype: object</code></pre>
<p>Uh oh. It looks like there&#8217;s something wrong with the profits column — we would expect it to be a <code>float64</code> like the revenue column. This indicates that it probably contains some non-integer values, so let&#8217;s take a look.</p>
<pre><code class="language-python">
non_numberic_profits = df.profit.str.contains('[^0-9.-]')
df.loc[non_numberic_profits].head()
</code></pre>
<div style="max-width: 100%;">
<style scoped="">
    .dataframe tbody tr th:only-of-type { vertical-align: middle; }</p>
<pre><code>.dataframe tbody tr th { vertical-align: top;}.dataframe thead th { text-align: right;}</code></pre>
</style>
<table border="1" class="dataframe">
<thead>
<tr style="text-align: right;">
<th>
        </th>
<th>
          year
        </th>
<th>
          rank
        </th>
<th>
          company
        </th>
<th>
          revenue
        </th>
<th>
          profit
        </th>
</tr>
</thead>
<tbody>
<tr>
<th>
          228
        </th>
<td>
          1955
        </td>
<td>
          229
        </td>
<td>
          Norton
        </td>
<td>
          135.0
        </td>
<td>
          N.A.
        </td>
</tr>
<tr>
<th>
          290
        </th>
<td>
          1955
        </td>
<td>
          291
        </td>
<td>
          Schlitz Brewing
        </td>
<td>
          100.0
        </td>
<td>
          N.A.
        </td>
</tr>
<tr>
<th>
          294
        </th>
<td>
          1955
        </td>
<td>
          295
        </td>
<td>
          Pacific Vegetable Oil
        </td>
<td>
          97.9
        </td>
<td>
          N.A.
        </td>
</tr>
<tr>
<th>
          296
        </th>
<td>
          1955
        </td>
<td>
          297
        </td>
<td>
          Liebmann Breweries
        </td>
<td>
          96.0
        </td>
<td>
          N.A.
        </td>
</tr>
<tr>
<th>
          352
        </th>
<td>
          1955
        </td>
<td>
          353
        </td>
<td>
          Minneapolis-Moline
        </td>
<td>
          77.4
        </td>
<td>
          N.A.
        </td>
</tr>
</tbody>
</table>
</div>
<p>Just as we suspected! Some of the values are strings, which have been used to indicate missing data. Are there any other values that have crept in?</p>
<pre class="language-python"><code class="language-python">set(df.profit[non_numberic_profits])</code></pre>
<pre class="language-python"><code class="language-python">{'N.A.'}</code></pre>
<p>That makes it easy to interpret, but what should we do? Well, that depends how many values are missing.</p>
<pre class="language-python"><code class="language-python">len(df.profit[non_numberic_profits])</code></pre>
<pre class="language-python"><code class="language-python">369</code></pre>
<p>It&#8217;s a small fraction of our data set, though not completely inconsequential as it is still around 1.5%. If rows containing <code>N.A.</code> are, roughly, uniformly distributed over the years, the easiest solution would just be to remove them. So let&#8217;s have a quick look at the distribution.</p>
<pre class="language-python"><code class="language-python">bin_sizes, _, _ = plt.hist(df.year[non_numberic_profits], bins=range(1955, 2006))</code></pre>
<p><noscript><img src="/wp-content/uploads/2019/01/jupyter-notebook-tutorial_35_0.png" alt="Missing value distribution" /></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="/wp-content/uploads/2019/01/jupyter-notebook-tutorial_35_0.png" alt="Missing value distribution" /></p>
<p>At a glance, we can see that the most invalid values in a single year is fewer than 25, and as there are 500 data points per year, removing these values would account for less than 4% of the data for the worst years. Indeed, other than a surge around the 90s, most years have fewer than half the missing values of the peak. For our purposes, let&#8217;s say this is acceptable and go ahead and remove these rows.</p>
<pre class="language-python"><code class="language-python">
df = df.loc[~non_numberic_profits]
df.profit = df.profit.apply(pd.to_numeric)
</code></pre>
<p>We should check that worked.</p>
<pre class="language-python"><code class="language-python">len(df)</code></pre>
<pre class="language-python"><code class="language-python">25131</code></pre>
<pre class="language-python"><code class="language-python">df.dtypes</code></pre>
<pre class="language-python"><code class="language-python">
year int64
rank int64
company object
revenue float64
profit float64
dtype: object</code></pre>
<p>Great! We have finished our data set setup.</p>
<p>If you were going to present your notebook as a report, you could get rid of the investigatory cells we created, which are included here as a demonstration of the flow of working with notebooks, and merge relevant cells (see the Advanced Functionality section below for more on this) to create a single data set setup cell. This would mean that if we ever mess up our data set elsewhere, we can just rerun the setup cell to restore it.</p>
<h3 id="plottingwithmatplotlib">Plotting with matplotlib</h3>
<p>Next, we can get to addressing the question at hand by plotting the average profit by year. We might as well plot the revenue as well, so first we can define some variables and a method to reduce our code.</p>
<pre><code class="language-python">
group_by_year = df.loc[:, ['year', 'revenue', 'profit']].groupby('year')
avgs = group_by_year.mean()
x = avgs.index
y1 = avgs.profit
def plot(x, y, ax, title, y_label):
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.plot(x, y)
    ax.margins(x=0, y=0)
</code></pre>
<p>Now let&#8217;s plot!</p>
<pre><code class="language-python">
fig, ax = plt.subplots()
plot(x, y1, ax, 'Increase in mean Fortune 500 company profits from 1955 to 2005', 'Profit (millions)')
</code></pre>
<p><noscript><img src="/wp-content/uploads/2019/01/jupyter-notebook-tutorial_44_0.png" alt="Increase in mean Fortune 500 company profits from 1955 to 2005" /></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="/wp-content/uploads/2019/01/jupyter-notebook-tutorial_44_0.png" alt="Increase in mean Fortune 500 company profits from 1955 to 2005" /></p>
<p>Wow, that looks like an exponential, but it&#8217;s got some huge dips. They must correspond to the <a href="https://en.wikipedia.org/wiki/Early_1990s_recession">early 1990s recession</a> and the <a href="https://en.wikipedia.org/wiki/Dot-com_bubble">dot-com bubble</a>. It&#8217;s pretty interesting to see that in the data. But how come profits recovered to even higher levels post each recession?</p>
<p>Maybe the revenues can tell us more.</p>
<pre><code class="language-python">
y2 = avgs.revenue
fig, ax = plt.subplots()
plot(x, y2, ax, 'Increase in mean Fortune 500 company revenues from 1955 to 2005', 'Revenue (millions)')
</code></pre>
<p><noscript><img src="/wp-content/uploads/2019/01/jupyter-notebook-tutorial_46_0.png" alt="Increase in mean Fortune 500 company revenues from 1955 to 2005" /></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="/wp-content/uploads/2019/01/jupyter-notebook-tutorial_46_0.png" alt="Increase in mean Fortune 500 company revenues from 1955 to 2005" /></p>
<p>That adds another side to the story. Revenues were no way nearly as badly hit, that&#8217;s some great accounting work from the finance departments.</p>
<p>With a little help <a href="https://stackoverflow.com/a/47582329/604687">from Stack Overflow</a>, we can superimpose these plots with +/- their standard deviations.</p>
<pre><code class="language-python">
def plot_with_std(x, y, stds, ax, title, y_label):
    ax.fill_between(x, y - stds, y + stds, alpha=0.2)
    plot(x, y, ax, title, y_label)
fig, (ax1, ax2) = plt.subplots(ncols=2)
title = 'Increase in mean and std Fortune 500 company %s from 1955 to 2005'
stds1 = group_by_year.std().profit.values
stds2 = group_by_year.std().revenue.values
plot_with_std(x, y1.values, stds1, ax1, title % 'profits', 'Profit (millions)')
plot_with_std(x, y2.values, stds2, ax2, title % 'revenues', 'Revenue (millions)')
fig.set_size_inches(14, 4)
fig.tight_layout()</code></pre>
<p><noscript><img src="https://www.dataquest.io/wp-content/uploads/2018/03/jupyter-notebook-tutorial_48_0.png" alt="jupyter-notebook-tutorial_48_0" /></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="https://www.dataquest.io/wp-content/uploads/2018/03/jupyter-notebook-tutorial_48_0.png" alt="jupyter-notebook-tutorial_48_0" /></p>
<p>That&#8217;s staggering, the standard deviations are huge. Some Fortune 500 companies make billions while others lose billions, and the risk has increased along with rising profits over the years. Perhaps some companies perform better than others; are the profits of the top 10% more or less volatile than the bottom 10%?</p>
<p>There are plenty of questions that we could look into next, and it&#8217;s easy to see how the flow of working in a notebook matches one&#8217;s own thought process, so now it&#8217;s time to draw this example to a close. This flow helped us to easily investigate our data set in one place without context switching between applications, and our work is immediately sharable and reproducible. If we wished to create a more concise report for a particular audience, we could quickly refactor our work by merging cells and removing intermediary code.</p>
<p><em>Note: The original version of this article used <code>as_matrix()</code> instead of <code>.values</code> in the code snippet above. As of this writing, <code>.as_matrix()</code> still works, but it is slated for removal in a future version of pandas, so we&#8217;ve replaced it with the future-proof <code>.values</code>.</em></p>
<h2 id="sharingyournotebooks">Sharing Your Notebooks</h2>
<p>When people talk of sharing their notebooks, there are generally two paradigms they may be considering. Most often, individuals share the end-result of their work, much like this article itself, which means sharing non-interactive, pre-rendered versions of their notebooks; however, it is also possible to collaborate on notebooks with the aid version control systems such as <a href="https://git-scm.com/">Git</a>.</p>
<p>That said, there are <a href="https://mybinder.org/">some</a> nascent <a href="https://kyso.io/">companies</a> popping up on the web offering the ability to run interactive Jupyter Notebooks in the cloud.</p>
<h3 id="beforeyoushare">Before You Share</h3>
<p>A shared notebook will appear exactly in the state it was in when you export or save it, including the output of any code cells. Therefore, to ensure that your notebook is share-ready, so to speak, there are a few steps you should take before sharing:</p>
<ol>
<li>Click &#8220;Cell > All Output > Clear&#8221;</li>
<li>Click &#8220;Kernel > Restart &amp; Run All&#8221;</li>
<li>Wait for your code cells to finish executing and check they did so as expected</li>
</ol>
<p>This will ensure your notebooks don&#8217;t contain intermediary output, have a stale state, and executed in order at the time of sharing.</p>
<h3 id="exportingyournotebooks">Exporting Your Notebooks</h3>
<p>Jupyter has built-in support for exporting to HTML and PDF as well as several other formats, which you can find from the menu under &#8220;File > Download As.&#8221; If you wish to share your notebooks with a small private group, this functionality may well be all you need. Indeed, as many researchers in academic institutions are given some public or internal webspace, and because you can export a notebook to an HTML file, Jupyter Notebooks can be an especially convenient way for them to share their results with their peers.</p>
<p>But if sharing exported files doesn&#8217;t cut it for you, there are also some immensely popular methods of sharing <code>.ipynb</code> files more directly on the web.</p>
<h3 id="github">GitHub</h3>
<p>With the <a href="https://github.com/parente/nbestimate">number of public notebooks on GitHub</a> exceeding 1.8 million by early 2018, it is surely the most popular independent platform for sharing Jupyter projects with the world. GitHub has integrated support for rendering <code>.ipynb</code> files directly both in repositories and gists on its website. If you aren&#8217;t already aware, <a href="https://github.com">GitHub</a> is a code hosting platform for version control and collaboration for repositories created with <a href="https://git-scm.com/">Git</a>. You&#8217;ll need an account to use their services, but standard accounts are free.</p>
<p>Once you have a GitHub account, the easiest way to share a notebook on GitHub doesn&#8217;t actually require Git at all. Since 2008, GitHub has provided its Gist service for hosting and sharing code snippets, which each get their own repository. To share a notebook using Gists:</p>
<ol>
<li>Sign in and browse to <a href="https://gist.github.com">gist.github.com</a>.</li>
<li>Open your <code>.ipynb</code> file in a text editor, select all and copy the JSON inside.</li>
<li>Paste the notebook JSON into the gist.</li>
<li>Give your Gist a filename, remembering to add <code>.iypnb</code> or this will not work.</li>
<li>Click either &#8220;Create secret gist&#8221; or &#8220;Create public gist.&#8221;</li>
</ol>
<p>This should look something like the following:</p>
<p><noscript><img src="/wp-content/uploads/2019/01/create-gist.jpg" alt="Creating a Gist" /></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="/wp-content/uploads/2019/01/create-gist.jpg" alt="Creating a Gist" /></p>
<p>If you created a public Gist, you will now be able to share its URL with anyone, and others will be able to <a href="https://help.github.com/articles/forking-and-cloning-gists/">fork and clone</a> your work.</p>
<p>Creating your own Git repository and sharing this on GitHub is beyond the scope of this tutorial, but <a href="https://guides.github.com/">GitHub provides plenty of guides</a> for you to get started on your own.</p>
<p>An extra tip for those using git is to <a href="https://stackoverflow.com/q/35916658/604687">add an exception</a> to your <code>.gitignore</code> for those hidden <code>.ipynb_checkpoints</code> directories Jupyter creates, so as not to commit checkpoint files unnecessarily to your repo.</p>
<h3 id="nbviewer">Nbviewer</h3>
<p>Having grown to render <a href="https://blog.jupyter.org/rendering-notebooks-on-github-f7ac8736d686">hundreds of thousands</a> of notebooks every week by 2015, NBViewer is the most popular notebook renderer on the web. If you already have somewhere to host your Jupyter Notebooks online, be it GitHub or elsewhere, NBViewer will render your notebook and provide a shareable URL along with it. Provided as a free service as part of Project Jupyter, it is available at <a href="https://nbviewer.jupyter.org/">nbviewer.jupyter.org</a>.</p>
<p>Initially developed before GitHub&#8217;s Jupyter Notebook integration, NBViewer allows anyone to enter a URL, Gist ID, or GitHub username/repo/file and it will render the notebook as a webpage. A Gist&#8217;s ID is the unique number at the end of its URL; for example, the string of characters after the last backslash in <code>https://gist.github.com/username/50896401c23e0bf417e89cd57e89e1de</code>. If you enter a GitHub username or username/repo, you will see a minimal file browser that lets you explore a user&#8217;s repos and their contents.</p>
<p>The URL NBViewer displays when displaying a notebook is a constant based on the URL of the notebook it is rendering, so you can share this with anyone and it will work as long as the original files remain online — NBViewer doesn&#8217;t cache files for very long.</p>
<h2 id="finalthoughts">Final Thoughts</h2>
<p>Starting with the basics, we have come to grips with the natural workflow of Jupyter Notebooks, delved into IPython&#8217;s more advanced features, and finally learned how to share our work with friends, colleagues, and the world. And we accomplished all this from a notebook itself!</p>
<p>It should be clear how notebooks promote a productive working experience by reducing context switching and emulating a natural development of thoughts during a project. The power of Jupyter Notebooks should also be evident, and we covered plenty of leads to get you started exploring more advanced features in your own projects.</p>
<p>If you&#8217;d like further inspiration for your own Notebooks, Jupyter has put together <a href="https://github.com/jupyter/jupyter/wiki/A-gallery-of-interesting-Jupyter-Notebooks">a gallery of interesting Jupyter Notebooks</a> that you may find helpful and the <a href="https://nbviewer.jupyter.org/">Nbviewer homepage</a> links to some really fancy examples of quality notebooks.</p>
<h2>More Great Jupyter Notebooks Resources</h2>
<ul>
<li><a href="https://www.dataquest.io/blog/advanced-jupyter-notebooks-tutorial/">Advanced Jupyter Notebooks Tutorial</a> &#8211; Now that you&#8217;ve mastered the basics, become a Jupyter Notebooks pro with this advanced tutorial!</li>
<li><a href="https://www.dataquest.io/blog/jupyter-notebook-tips-tricks-shortcuts/">28 Jupyter Notebooks Tips, Tricks, and Shortcuts</a> &#8211; Make yourself into a power user and increase your efficiency with these tips and tricks!</li>
<li><a href="https://www.dataquest.io/m/349-project-learn-and-install-jupyter-notebook/">Guided Project &#8211; Install and Learn Jupyter Notebooks</a> &#8211; Give yourself a great foundation working with Jupyter Notebooks by working through this interactive guided project that&#8217;ll get you set up and teach you the ropes.</li>
</ul>
<p><em>This post was updated in August 2019.</em><br />
<span style="display:none" class="tl-placeholder-f-type-shortcode_19063"></span></p>
<span id="tho-end-content" style="display: block; visibility: hidden;"></span><div class="saboxplugin-wrap" itemtype="http://schema.org/Person" itemscope itemprop="author"><div class="saboxplugin-gravatar"><noscript><img src="https://www.dataquest.io/wp-content/uploads/2015/12/benjamin_pryke_profile-150x150.jpg" alt="Benjamin Pryke" class="avatar avatar-100 wp-user-avatar wp-user-avatar-100 alignnone photo" /></noscript><img src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20210%20140%22%3E%3C/svg%3E' data-src="https://www.dataquest.io/wp-content/uploads/2015/12/benjamin_pryke_profile-150x150.jpg" alt="Benjamin Pryke" class="lazyload avatar avatar-100 wp-user-avatar wp-user-avatar-100 alignnone photo" /></div><div class="saboxplugin-authorname"><a href="https://www.dataquest.io/blog/author/benjamin-pryke/" class="vcard author" rel="author" itemprop="url"><span class="fn" itemprop="name">Benjamin Pryke</span></a></div><div class="saboxplugin-desc"><div itemprop="description"><p>Python and web developer with a background in computer science and machine learning. Co-founder of FinTech firm Machina Capital. Part-time gymnast and digital bohemian.</p>
</div></div><div class="clearfix"></div></div></section>
<div class="thrv_wrapper thrv-divider theme-bottom-divider" data-style="tve_sep-1" data-thickness="1" data-color="rgb(225, 225, 225)" data-css="tve-u-16ec5fddeba" id="Divider" data-color-d="rgba(0, 0, 0, 0.06)" style="padding-bottom: 15px !important;">
<hr class="tve_sep tve_sep-1" style="">
</div><div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-16ec634b5e1"><p data-css="tve-u-16ec63322d0">Tags</p></div><div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-16ec6330c3d"><p data-css="tve-u-16ec63492a5"><span class="thrive-shortcode-content" data-attr-link="0" data-attr-rel="0" data-attr-target="1" data-css="tve-u-16ec635199f" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_tags" data-shortcode-name="Post Tags">beginner, Jupyter, jupyter notebooks, Learn Python, Pandas, python, tutorial, Tutorials</span></p></div><div class="thrv_wrapper thrv-divider theme-bottom-divider" data-style="tve_sep-1" data-thickness="1" data-color="rgb(225, 225, 225)" data-css="tve-u-16ecb61f672" data-color-d="rgba(0, 0, 0, 0.06)" style="">
<hr class="tve_sep tve_sep-1" style="">
</div><div class="thrv_wrapper thrv_text_element"><h4 class="" data-css="tve-u-170199ce799"><strong>You may also like</strong></h4></div><div class="tcb-post-list thrv_wrapper" data-type="grid" data-pagination-type="none" data-pages_near_current="2" data-css="tve-u-17146113c1f" data-total_post_count="298" data-total_sticky_count="0" data-disabled-links="1" data-no_posts_text=""><article id="post-14526" class="post-14526 post type-post status-publish format-standard has-post-thumbnail hentry category-data-science-career-tips tag-career tag-career-tips tag-certificates tag-study post-wrapper thrv_wrapper  tve_evt_manager_listen tve_ea_thrive_animation tve_et_mouseover tve_anim_grow" tcb_hover_state_parent="" class="tve_evt_manager_listen tve_ea_thrive_animation tve_et_mouseover tve_anim_grow" data-selector=".post-wrapper" data-permalink="https://www.dataquest.io/blog/data-science-certificate/"><style class="tcb-post-list-dynamic-style" type="text/css">@media (min-width: 300px){.tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-14526.post-wrapper{background-image: linear-gradient(rgba(0, 0, 0, 0.55), rgba(0, 0, 0, 0.55)), url("https://www.dataquest.io/wp-content/uploads/2019/04/data-science-certificate.jpg") !important;}:not(#tve) .tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-14526.post-wrapper:hover{background-image: linear-gradient(rgba(0, 0, 0, 0.5), rgba(0, 0, 0, 0.5)), url("https://www.dataquest.io/wp-content/uploads/2019/04/data-science-certificate.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-14526 [data-css="tve-u-16eb7cc1def"]{background-image: url("https://www.dataquest.io/wp-content/uploads/2019/04/data-science-certificate.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-14526 [data-css="tve-u-16eb7cc1dee"]:hover [data-css="tve-u-16eb7cc1def"]{background-image: linear-gradient(rgba(0, 0, 0, 0.05), rgba(0, 0, 0, 0.05)), url("https://www.dataquest.io/wp-content/uploads/2019/04/data-science-certificate.jpg") !important;}}</style>
		<div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-17146113c23" style="">
	<div class="tve-content-box-background" data-css="tve-u-17146113c24" data-tcb_hover_state_parent="1"></div>
	<div class="tve-cb" data-css="tve-u-17146113c26"><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-17146113c27">
	<div class="tve-content-box-background" data-css="tve-u-17146113c28" data-tcb_hover_state_parent="1"></div>
	<div class="tve-cb" data-css="tve-u-17146113c29"><div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-17146113c2a"><h2 class="" style="" data-css="tve-u-17146113c2b"><span class="thrive-shortcode-content" data-attr-link="1" data-attr-rel="0" data-attr-target="0" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_title" data-shortcode-name="Post title" data-css="tve-u-17146113c2c" style="" data-attr-css=""><a href="https://www.dataquest.io/blog/data-science-certificate/" title="Data Science Certificates in 2020 (Do You Really Need One?)" data-css="">Data Science Certificates in 2020 (Do You Really Need One?)</a></span></h2></div><div class="tcb-post-read-more thrv_wrapper" data-css="tve-u-17146113c2d" style="">
				<a href="https://www.dataquest.io/blog/data-science-certificate/" class="tcb-button-link tcb-post-read-more-link">
					<span class="tcb-button-texts">
						<span class="tcb-button-text thrv-inline-text">​Read More</span>
					</span>
				</a>
			</div></div>
</div></div>
</div>
		

		

		

		
	<div class="tve-article-cover"></div></article><article id="post-23820" class="post-23820 post type-post status-publish format-standard has-post-thumbnail hentry category-dataquest-updates category-student-stories tag-outcomes-survey tag-survey post-wrapper thrv_wrapper  tve_evt_manager_listen tve_ea_thrive_animation tve_et_mouseover tve_anim_grow" tcb_hover_state_parent="" class="tve_evt_manager_listen tve_ea_thrive_animation tve_et_mouseover tve_anim_grow" data-selector=".post-wrapper" data-permalink="https://www.dataquest.io/blog/is-dataquest-worth-it-2020/"><style class="tcb-post-list-dynamic-style" type="text/css">@media (min-width: 300px){.tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-23820.post-wrapper{background-image: linear-gradient(rgba(0, 0, 0, 0.55), rgba(0, 0, 0, 0.55)), url("https://www.dataquest.io/wp-content/uploads/2020/05/is-dataquest-worth-it-improve-life.jpg") !important;}:not(#tve) .tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-23820.post-wrapper:hover{background-image: linear-gradient(rgba(0, 0, 0, 0.5), rgba(0, 0, 0, 0.5)), url("https://www.dataquest.io/wp-content/uploads/2020/05/is-dataquest-worth-it-improve-life.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-23820 [data-css="tve-u-16eb7cc1def"]{background-image: url("https://www.dataquest.io/wp-content/uploads/2020/05/is-dataquest-worth-it-improve-life.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-23820 [data-css="tve-u-16eb7cc1dee"]:hover [data-css="tve-u-16eb7cc1def"]{background-image: linear-gradient(rgba(0, 0, 0, 0.05), rgba(0, 0, 0, 0.05)), url("https://www.dataquest.io/wp-content/uploads/2020/05/is-dataquest-worth-it-improve-life.jpg") !important;}}</style>
		<div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-17146113c23" style="">
	<div class="tve-content-box-background" data-css="tve-u-17146113c24" data-tcb_hover_state_parent="1"></div>
	<div class="tve-cb" data-css="tve-u-17146113c26"><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-17146113c27">
	<div class="tve-content-box-background" data-css="tve-u-17146113c28" data-tcb_hover_state_parent="1"></div>
	<div class="tve-cb" data-css="tve-u-17146113c29"><div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-17146113c2a"><h2 class="" style="" data-css="tve-u-17146113c2b"><span class="thrive-shortcode-content" data-attr-link="1" data-attr-rel="0" data-attr-target="0" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_title" data-shortcode-name="Post title" data-css="tve-u-17146113c2c" style="" data-attr-css=""><a href="https://www.dataquest.io/blog/is-dataquest-worth-it-2020/" title="Is Dataquest Worth It? Here’s What Our Learners Say in 2020" data-css="">Is Dataquest Worth It? Here’s What Our Learners Say in 2020</a></span></h2></div><div class="tcb-post-read-more thrv_wrapper" data-css="tve-u-17146113c2d" style="">
				<a href="https://www.dataquest.io/blog/is-dataquest-worth-it-2020/" class="tcb-button-link tcb-post-read-more-link">
					<span class="tcb-button-texts">
						<span class="tcb-button-text thrv-inline-text">​Read More</span>
					</span>
				</a>
			</div></div>
</div></div>
</div>
		

		

		

		
	<div class="tve-article-cover"></div></article><article id="post-1943" class="post-1943 post type-post status-publish format-standard has-post-thumbnail hentry category-dataquest-updates category-learning-and-motivation tag-guides tag-how-to tag-learning tag-motivation post-wrapper thrv_wrapper  tve_evt_manager_listen tve_ea_thrive_animation tve_et_mouseover tve_anim_grow" tcb_hover_state_parent="" class="tve_evt_manager_listen tve_ea_thrive_animation tve_et_mouseover tve_anim_grow" data-selector=".post-wrapper" data-permalink="https://www.dataquest.io/blog/learn-data-science/"><style class="tcb-post-list-dynamic-style" type="text/css">@media (min-width: 300px){.tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-1943.post-wrapper{background-image: linear-gradient(rgba(0, 0, 0, 0.55), rgba(0, 0, 0, 0.55)), url("https://www.dataquest.io/wp-content/uploads/2020/05/learn-data-science.jpg") !important;}:not(#tve) .tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-1943.post-wrapper:hover{background-image: linear-gradient(rgba(0, 0, 0, 0.5), rgba(0, 0, 0, 0.5)), url("https://www.dataquest.io/wp-content/uploads/2020/05/learn-data-science.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-1943 [data-css="tve-u-16eb7cc1def"]{background-image: url("https://www.dataquest.io/wp-content/uploads/2020/05/learn-data-science.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-1943 [data-css="tve-u-16eb7cc1dee"]:hover [data-css="tve-u-16eb7cc1def"]{background-image: linear-gradient(rgba(0, 0, 0, 0.05), rgba(0, 0, 0, 0.05)), url("https://www.dataquest.io/wp-content/uploads/2020/05/learn-data-science.jpg") !important;}}</style>
		<div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-17146113c23" style="">
	<div class="tve-content-box-background" data-css="tve-u-17146113c24" data-tcb_hover_state_parent="1"></div>
	<div class="tve-cb" data-css="tve-u-17146113c26"><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-17146113c27">
	<div class="tve-content-box-background" data-css="tve-u-17146113c28" data-tcb_hover_state_parent="1"></div>
	<div class="tve-cb" data-css="tve-u-17146113c29"><div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-17146113c2a"><h2 class="" style="" data-css="tve-u-17146113c2b"><span class="thrive-shortcode-content" data-attr-link="1" data-attr-rel="0" data-attr-target="0" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_title" data-shortcode-name="Post title" data-css="tve-u-17146113c2c" style="" data-attr-css=""><a href="https://www.dataquest.io/blog/learn-data-science/" title="How to Learn Data Science (Step-By-Step) in 2020" data-css="">How to Learn Data Science (Step-By-Step) in 2020</a></span></h2></div><div class="tcb-post-read-more thrv_wrapper" data-css="tve-u-17146113c2d" style="">
				<a href="https://www.dataquest.io/blog/learn-data-science/" class="tcb-button-link tcb-post-read-more-link">
					<span class="tcb-button-texts">
						<span class="tcb-button-text thrv-inline-text">​Read More</span>
					</span>
				</a>
			</div></div>
</div></div>
</div>
		

		

		

		
	<div class="tve-article-cover"></div></article><article id="post-23497" class="post-23497 post type-post status-publish format-standard has-post-thumbnail hentry category-dataquest-updates tag-announcements tag-course-launches post-wrapper thrv_wrapper  tve_evt_manager_listen tve_ea_thrive_animation tve_et_mouseover tve_anim_grow" tcb_hover_state_parent="" class="tve_evt_manager_listen tve_ea_thrive_animation tve_et_mouseover tve_anim_grow" data-selector=".post-wrapper" data-permalink="https://www.dataquest.io/blog/new-course-learn-machine-learning-in-r/"><style class="tcb-post-list-dynamic-style" type="text/css">@media (min-width: 300px){.tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-23497.post-wrapper{background-image: linear-gradient(rgba(0, 0, 0, 0.55), rgba(0, 0, 0, 0.55)), url("https://www.dataquest.io/wp-content/uploads/2020/04/learn-machine-learning-in-r.jpg") !important;}:not(#tve) .tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-23497.post-wrapper:hover{background-image: linear-gradient(rgba(0, 0, 0, 0.5), rgba(0, 0, 0, 0.5)), url("https://www.dataquest.io/wp-content/uploads/2020/04/learn-machine-learning-in-r.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-23497 [data-css="tve-u-16eb7cc1def"]{background-image: url("https://www.dataquest.io/wp-content/uploads/2020/04/learn-machine-learning-in-r.jpg") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-23497 [data-css="tve-u-16eb7cc1dee"]:hover [data-css="tve-u-16eb7cc1def"]{background-image: linear-gradient(rgba(0, 0, 0, 0.05), rgba(0, 0, 0, 0.05)), url("https://www.dataquest.io/wp-content/uploads/2020/04/learn-machine-learning-in-r.jpg") !important;}}</style>
		<div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-17146113c23" style="">
	<div class="tve-content-box-background" data-css="tve-u-17146113c24" data-tcb_hover_state_parent="1"></div>
	<div class="tve-cb" data-css="tve-u-17146113c26"><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-17146113c27">
	<div class="tve-content-box-background" data-css="tve-u-17146113c28" data-tcb_hover_state_parent="1"></div>
	<div class="tve-cb" data-css="tve-u-17146113c29"><div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-17146113c2a"><h2 class="" style="" data-css="tve-u-17146113c2b"><span class="thrive-shortcode-content" data-attr-link="1" data-attr-rel="0" data-attr-target="0" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_title" data-shortcode-name="Post title" data-css="tve-u-17146113c2c" style="" data-attr-css=""><a href="https://www.dataquest.io/blog/new-course-learn-machine-learning-in-r/" title="New Course: Introduction to Machine Learning in R" data-css="">New Course: Introduction to Machine Learning in R</a></span></h2></div><div class="tcb-post-read-more thrv_wrapper" data-css="tve-u-17146113c2d" style="">
				<a href="https://www.dataquest.io/blog/new-course-learn-machine-learning-in-r/" class="tcb-button-link tcb-post-read-more-link">
					<span class="tcb-button-texts">
						<span class="tcb-button-text thrv-inline-text">​Read More</span>
					</span>
				</a>
			</div></div>
</div></div>
</div>
		

		

		

		
	<div class="tve-article-cover"></div></article></div>
</div></div>
</div>
<div class="main-content-background"></div></div><div id="theme-bottom-section" class="theme-section bottom-section"><div class="section-background"></div><div class="section-content"><div class="thrv_wrapper thrv-page-section tcb-window-width" data-inherit-lp-settings="1" style="width: 1140px; left: -30.015px;" data-css="tve-u-1714ac8605e">
	<div class="tve-page-section-out" style="" data-css="tve-u-1714ac8605a"></div>
	<div class="tve-page-section-in tve_empty_dropzone" data-css="tve-u-1714ac86059" style=""><div class="thrv_wrapper thrv_icon tcb-icon-display" data-css="tve-u-1714ac8606b" style=""><svg class="tcb-icon" viewBox="0 0 512 512" data-id="icon-rocket-light" data-name="" style="">
            <path d="M505.09 19.34A16.08 16.08 0 0 0 492.73 7c-32.61-7-58.13-7-83.56-7C305.44 0 243 55.1 196.27 128H94.87A48 48 0 0 0 52 154.49l-49.42 98.8A24 24 0 0 0 24.07 288h92.6l-10.59 21.42a31.75 31.75 0 0 0 6.15 36.63L166 399.77a32 32 0 0 0 22.61 9.43 31.58 31.58 0 0 0 14-3.29L224 395.33V488a24.08 24.08 0 0 0 24 24 23.66 23.66 0 0 0 10.63-2.53l98.7-49.39a47.91 47.91 0 0 0 26.5-42.9V315.71C456.61 268.93 512 206.25 512 103c.08-25.53.08-51-6.91-83.66zM128 256H37l43.59-87.2a16.09 16.09 0 0 1 14.28-8.8h85.18c-13.94 28.07-31.72 64-47.56 96zm223.86 161.18a16 16 0 0 1-8.86 14.28L256 475v-95.52c32-15.83 67.82-33.59 95.86-47.54zM366.5 288.9c-39.5 19.78-135.87 67.45-177.91 88.24l-53.82-53.55C155.64 281.36 203.3 185 223 145.41 273.21 67 330.68 32 409.17 32c21.54 0 42 0 66.63 4.29 4.34 24.87 4.26 45.21 4.18 66.7C480 181.07 445 238.46 366.5 288.9zM368 80a64 64 0 1 0 64 64 64.07 64.07 0 0 0-64-64zm0 96a32 32 0 1 1 32-32 32 32 0 0 1-32 32z"></path>
        </svg></div><div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-1714ac8605d"><h2 class="" style="text-align: center;"><span data-css="tve-u-1714ac86068">Get started with Dataquest today -</span>
<span data-css="tve-u-1714ac8605c">for free!</span></h2></div><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box" data-css="tve-u-1714ac86064" style="">
	<div class="tve-content-box-background" data-css="tve-u-1714ac86061" style=""></div>
	<div class="tve-cb" data-css="tve-u-1714ac86063"><div class="thrv_wrapper thrv_custom_html_shortcode" data-css="tve-u-1717d14bf10" style=""><iframe src="https://app.dataquest.io/signup#iframe" style="padding: 0px; margin: 0px; width: 100%; min-height: 320px; border: 0px; " border="0"></div></iframe><div class="tve_iframe_cover"></div></div><div class="thrv_wrapper thrv_text_element tve-froala fr-box fr-basic" data-css="tve-u-1714ac86066"><p data-css="tve-u-1714ac86067" style="text-align: center;">By creating an account you agree to accept our <a class="tve-froala" href="/terms/">terms of use</a>&nbsp;and&nbsp;<a class="tve-froala" href="/privacy/" style="outline: none;">privacy policy</a>.</p></div></div>
</div><div class="thrv_wrapper thrv_text_element tve-froala fr-box fr-basic" style="" data-css="tve-u-1714ac86060"><p data-css="tve-u-1714ac8605f" style="text-align: center;">​<a class="tve-froala fr-basic" data-css="tve-u-1714ac86069" href="https://www.dataquest.io/subscribe-test/">Or, visit our pricing page to learn about our Basic and Premium plans.</a></p></div></div>
</div></div></div><footer id="thrive-footer" class="thrv_wrapper thrv_symbol thrv_footer thrv_symbol_23142" role="contentinfo"><div class="thrive-shortcode-html thrive-symbol-shortcode" data-symbol-id="23142"><style class='tve-symbol-custom-style'>@media (min-width: 300px){.thrv_symbol_23142 .symbol-section-out { background-color: rgb(38, 53, 85)  !important; }.thrv_symbol_23142 .symbol-section-in { --tve-font-size:16px; padding: 20px !important; min-height: 1px !important; }.thrv_symbol_23142 [data-css="tve-u-1705da4f349"] { width: 183px; margin-top: 0px !important; margin-bottom: 15px !important; }.thrv_symbol_23142 .thrive-symbol-shortcode { border-top: none; padding: 20px !important; }:not(#tve) .thrv_symbol_23142 .symbol-section-in p, :not(#tve) .thrv_symbol_23142 .symbol-section-in li, :not(#tve) .thrv_symbol_23142 .symbol-section-in blockquote, :not(#tve) .thrv_symbol_23142 .symbol-section-in address, :not(#tve) .thrv_symbol_23142 .symbol-section-in .tcb-plain-text, :not(#tve) .thrv_symbol_23142 .symbol-section-in label { font-size: var(--tve-font-size, 16px); }.thrv_symbol_23142 [data-css="tve-u-1705dcffc85"]::after { clear: both; }.thrv_symbol_23142 [data-css="tve-u-17062d767d9"] { padding: 30px 0px !important; }.thrv_symbol_23142 [data-css="tve-u-17062d76b70"] { border-top: none; margin: 0px !important; }.thrv_symbol_23142 [data-css="tve-u-17062d824d3"] { max-width: 60.4%; }.thrv_symbol_23142 [data-css="tve-u-17062d82581"] { max-width: 39.6%; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-17145d71f96"] { color: rgb(255, 255, 255)  !important; --tcb-applied-color:var$(--tcb-color-12)  !important; font-size: 12px !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-17145d71f99"] { color: rgb(255, 255, 255)  !important; --tcb-applied-color:var$(--tcb-color-12)  !important; font-size: 12px !important; }.thrv_symbol_23142 [data-css="tve-u-17145d87d5b"] { max-width: 500px; padding-right: 30px !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-17145d8fc5b"] { color: rgb(255, 255, 255)  !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-17145d8fc5b"]:hover { color: rgb(255, 255, 255)  !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-17147ee92c5"] { font-size: 12px !important; }.thrv_symbol_23142 [data-css="tve-u-1714808b5a7"] { color: rgb(255, 255, 255)  !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-17148090d4b"] { color: rgb(255, 255, 255)  !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-1714812ccf9"] { color: rgb(255, 255, 255)  !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-1714815996d"] { font-size: 11px !important; line-height: 2em !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-171481697a4"] { color: rgb(255, 255, 255)  !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-1714816a935"] { color: rgb(255, 255, 255)  !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-17148199b19"] { color: rgb(255, 255, 255)  !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-171481be72e"] { color: rgb(255, 255, 255)  !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-171481c3bd2"] { color: rgb(255, 255, 255)  !important; }.thrv_symbol_23142 [data-css="tve-u-171481dcc4c"] { max-width: 40.4%; }.thrv_symbol_23142 [data-css="tve-u-171481dccf1"] { max-width: 30.6%; }.thrv_symbol_23142 [data-css="tve-u-171481dcd94"] { max-width: 29%; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-1714f61ae0a"] { font-size: 11px !important; line-height: 2em !important; color: rgb(255, 255, 255)  !important; --tcb-applied-color:var$(--tcb-color-12)  !important; }.thrv_symbol_23142 [data-css="tve-u-1714f61ae45"] { color: rgb(255, 255, 255)  !important; }:not(#tve) .thrv_symbol_23142 [data-css="tve-u-1714f61ceed"] { color: rgb(255, 255, 255)  !important; }}@media (max-width: 1023px){.thrv_symbol_23142 .thrive-symbol-shortcode { padding-top: 20px !important; padding-left: 20px !important; }}@media (max-width: 767px){.thrv_symbol_23142 [data-css="tve-u-1705da4f349"] { float: left; z-index: 3; position: relative; width: 140px; margin-bottom: 30px !important; }.thrv_symbol_23142 [data-css="tve-u-1705dcffc85"]::after { clear: both; }.thrv_symbol_23142 .thrive-symbol-shortcode { padding-top: 20px !important; padding-left: 20px !important; padding-right: 20px !important; }}</style><div class="symbol-section-out" style=""></div> <div class="symbol-section-in" style=""> <div class="thrv_wrapper thrv-columns" style="" data-css="tve-u-17062d76b70"><div class="tcb-flex-row tcb--cols--2 tcb-resized" style="" data-css="tve-u-17062d767d9"><div class="tcb-flex-col" data-css="tve-u-17062d824d3" style=""><div class="tcb-col"><div class="tcb-clear" data-css="tve-u-1705dcffc85"><a class="tcb-logo thrv_wrapper" href="https://www.dataquest.io" data-css="tve-u-1705da4f349" class=""><picture><source data-srcset="https://www.dataquest.io/wp-content/uploads/2020/04/dataquest_logo_whited120f9c4e052e730254ee717e636361b-300x53.png 300w, https://www.dataquest.io/wp-content/uploads/2020/04/dataquest_logo_whited120f9c4e052e730254ee717e636361b.png 476w" media="(min-width:1024px)"></source><noscript><img src="https://www.dataquest.io/wp-content/uploads/2020/04/dataquest_logo_whited120f9c4e052e730254ee717e636361b.png" height="84" width="476" alt="" loading="lazy"></noscript><img class="lazyload" src='data:image/svg+xml,%3Csvg%20xmlns=%22http://www.w3.org/2000/svg%22%20viewBox=%220%200%20476%2084%22%3E%3C/svg%3E' data-src="https://www.dataquest.io/wp-content/uploads/2020/04/dataquest_logo_whited120f9c4e052e730254ee717e636361b.png" height="84" width="476" alt="" loading="lazy"></img></picture></a></div><div class="thrv_wrapper thrv_text_element tve-froala fr-box fr-basic" style="" data-css="tve-u-17145d87d5b"><p data-css="tve-u-17145d71f96">All rights reserved © 2019 – Dataquest Labs, Inc.</p><p data-css="tve-u-17145d71f99">We are committed to protecting your personal information and your right to privacy. Privacy Policy last updated June 13th, 2019 –&nbsp;<a class="tve-froala fr-basic" data-css="tve-u-17145d8fc5b" data-tcb_hover_state_parent="" href="https://www.dataquest.io/privacy" style="outline: none;">review here</a>.</p></div></div></div><div class="tcb-flex-col" data-css="tve-u-17062d82581" style=""><div class="tcb-col"><div class="thrv_wrapper thrv-columns"><div class="tcb-flex-row v-2 tcb--cols--3 tcb-resized"><div class="tcb-flex-col" data-css="tve-u-171481dcc4c" style=""><div class="tcb-col"><div class="thrv_wrapper thrv_text_element tve-froala fr-box fr-basic"><div class="tcb-plain-text" data-css="tve-u-1714815996d"><span data-css="tve-u-1714808b5a7" style="color: var(--tcb-color-12);"><a class="tve-froala fr-basic" href="/directory/" style="outline: none;">Data Science Courses&nbsp;</a></span><br><a class="tve-froala fr-basic" data-css="tve-u-17148090d4b" href="/for-business/" style="outline: none;">For Business</a><br><a class="tve-froala fr-basic" data-css="tve-u-1714812ccf9" href="/for-institutions/" style="outline: none;">For Academia</a><br><a class="tve-froala fr-basic" data-css="tve-u-1714816a935" href="/were-hiring/" style="outline: none;" target="_blank">We're Hiring<br></a><a class="tve-froala fr-basic" data-css="tve-u-171481697a4" href="/blog/topics/student-stories/" style="outline: none;" target="_blank">Success Stories</a></div><div class="tcb-plain-text" data-css="tve-u-17147ee92c5"><br><br></div></div></div></div><div class="tcb-flex-col" data-css="tve-u-171481dccf1" style=""><div class="tcb-col"><div class="thrv_wrapper thrv_text_element tve-froala fr-box fr-basic"><div class="tcb-plain-text" data-css="tve-u-1714f61ae0a"><a class="tve-froala fr-basic" data-css="tve-u-1714f61ceed" href="/subscribe/" style="outline: none;">Pricing</a></div><div class="tcb-plain-text" data-css="tve-u-1714815996d"><span data-css="tve-u-1714f61ae45" style="color: var(--tcb-color-12);"><a class="tve-froala fr-basic" href="https://support.dataquest.io/portal/signin" style="outline: none;" target="_blank">Contact Us</a></span><br><a class="tve-froala fr-basic" data-css="tve-u-17148090d4b" href="https://community.dataquest.io/" style="outline: none;" target="_blank">Community</a><br><a class="tve-froala fr-basic" data-css="tve-u-1714812ccf9" href="/privacy/" style="outline: none;">Privacy Policy</a><br><a class="tve-froala fr-basic" data-css="tve-u-17148199b19" href="/terms/" style="outline: none;">Terms of Use<br></a><a class="tve-froala fr-basic show-placeholder" data-css="tve-u-1714816a935" href="/were-hiring/" style="outline: none;" target="_blank">​</a><a class="tve-froala fr-basic show-placeholder" data-css="tve-u-171481697a4" href="/refund-policy/" style="outline: none;">​</a></div><div class="tcb-plain-text" data-css="tve-u-17147ee92c5"><br><br></div></div></div></div><div class="tcb-flex-col" data-css="tve-u-171481dcd94" style=""><div class="tcb-col"><div class="thrv_wrapper thrv_text_element tve-froala fr-box fr-basic"><div class="tcb-plain-text" data-css="tve-u-1714815996d"><span data-css="tve-u-1714808b5a7" style="color: var(--tcb-color-12);"><a class="tve-froala fr-basic" href="/blog/" style="outline: none;">Blog</a></span><br><a class="tve-froala fr-basic" data-css="tve-u-17148090d4b" href="https://www.facebook.com/dataquestio" style="outline: none;" target="_blank">Facebook</a><br><a class="tve-froala fr-basic" data-css="tve-u-1714812ccf9" href="https://twitter.com/dataquestio" style="outline: none;" target="_blank">Twitter</a></div><div class="tcb-plain-text" data-css="tve-u-1714815996d"><a class="tve-froala fr-basic" data-css="tve-u-171481be72e" href="https://www.linkedin.com/company/dataquest-io" style="outline: none;" target="_blank">LinkedIn</a><br><a class="tve-froala fr-basic" data-css="tve-u-171481c3bd2" href="/data-science-resources/" style="outline: none;">Resource List</a><br><a class="tve-froala fr-basic show-placeholder" data-css="tve-u-1714816a935" href="https://www.linkedin.com/company/dataquest-io" style="outline: none;" target="_blank">​</a></div><div class="tcb-plain-text" data-css="tve-u-17147ee92c5"><br><br></div></div></div></div></div></div></div></div></div></div> </div><style type="text/css" class="tve_global_style"></style></div></footer></div>
<script type='text/javascript'>
/* <![CDATA[ */
var TVE_Ult_Data = {"ajaxurl":"https:\/\/www.dataquest.io\/wp-admin\/admin-ajax.php","ajax_load_action":"tve_ult_ajax_load","conversion_events_action":"tve_ult_conversion_event","shortcode_campaign_ids":[],"matched_display_settings":[],"campaign_ids":[],"post_id":2024,"is_singular":true,"tu_em":""};
/* ]]> */
</script>
		<script type="text/javascript" src="https://www.dataquest.io/wp-content/plugins/thrive-ultimatum/js/dist/no-campaign.min.js?v=2.2.13.3"></script><span style="display:none" class="tl-placeholder-f-type-ribbon"></span><style>.lazyload,.lazyloading{opacity:0;}.lazyloaded{opacity:1;transition:opacity 300ms;}</style><noscript><style>.lazyload{display:none;}</style></noscript><script data-noptimize="1">window.lazySizesConfig=window.lazySizesConfig||{};window.lazySizesConfig.loadMode=1;</script><script async data-noptimize="1" src='https://www.dataquest.io/wp-content/plugins/autoptimize/classes/external/js/lazysizes.min.js?ao_version=2.7.1'></script><style type="text/css">.saboxplugin-wrap{-webkit-box-sizing:border-box;-moz-box-sizing:border-box;-ms-box-sizing:border-box;box-sizing:border-box;border:1px solid #eee;width:100%;clear:both;display:block;overflow:hidden;word-wrap:break-word;position:relative}.saboxplugin-wrap .saboxplugin-gravatar{float:left;padding:20px}.saboxplugin-wrap .saboxplugin-gravatar img{max-width:100px;height:auto;border-radius:0;}.saboxplugin-wrap .saboxplugin-authorname{font-size:18px;line-height:1;margin:20px 0 0 20px;display:block}.saboxplugin-wrap .saboxplugin-authorname a{text-decoration:none}.saboxplugin-wrap .saboxplugin-authorname a:focus{outline:0}.saboxplugin-wrap .saboxplugin-desc{display:block;margin:5px 20px}.saboxplugin-wrap .saboxplugin-desc a{text-decoration:underline}.saboxplugin-wrap .saboxplugin-desc p{margin:5px 0 12px}.saboxplugin-wrap .saboxplugin-web{margin:0 20px 15px;text-align:left}.saboxplugin-wrap .sab-web-position{text-align:right}.saboxplugin-wrap .saboxplugin-web a{color:#ccc;text-decoration:none}.saboxplugin-wrap .saboxplugin-socials{position:relative;display:block;background:#fcfcfc;padding:5px;border-top:1px solid #eee}.saboxplugin-wrap .saboxplugin-socials a svg{width:20px;height:20px}.saboxplugin-wrap .saboxplugin-socials a svg .st2{fill:#fff; transform-origin:center center;}.saboxplugin-wrap .saboxplugin-socials a svg .st1{fill:rgba(0,0,0,.3)}.saboxplugin-wrap .saboxplugin-socials a:hover{opacity:.8;-webkit-transition:opacity .4s;-moz-transition:opacity .4s;-o-transition:opacity .4s;transition:opacity .4s;box-shadow:none!important;-webkit-box-shadow:none!important}.saboxplugin-wrap .saboxplugin-socials .saboxplugin-icon-color{box-shadow:none;padding:0;border:0;-webkit-transition:opacity .4s;-moz-transition:opacity .4s;-o-transition:opacity .4s;transition:opacity .4s;display:inline-block;color:#fff;font-size:0;text-decoration:inherit;margin:5px;-webkit-border-radius:0;-moz-border-radius:0;-ms-border-radius:0;-o-border-radius:0;border-radius:0;overflow:hidden}.saboxplugin-wrap .saboxplugin-socials .saboxplugin-icon-grey{text-decoration:inherit;box-shadow:none;position:relative;display:-moz-inline-stack;display:inline-block;vertical-align:middle;zoom:1;margin:10px 5px;color:#444;fill:#444}.clearfix:after,.clearfix:before{content:' ';display:table;line-height:0;clear:both}.ie7 .clearfix{zoom:1}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-twitch{border-color:#38245c}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-addthis{border-color:#e91c00}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-behance{border-color:#003eb0}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-delicious{border-color:#06c}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-deviantart{border-color:#036824}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-digg{border-color:#00327c}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-dribbble{border-color:#ba1655}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-facebook{border-color:#1e2e4f}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-flickr{border-color:#003576}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-github{border-color:#264874}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-google{border-color:#0b51c5}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-googleplus{border-color:#96271a}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-html5{border-color:#902e13}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-instagram{border-color:#1630aa}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-linkedin{border-color:#00344f}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-pinterest{border-color:#5b040e}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-reddit{border-color:#992900}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-rss{border-color:#a43b0a}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-sharethis{border-color:#5d8420}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-skype{border-color:#00658a}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-soundcloud{border-color:#995200}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-spotify{border-color:#0f612c}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-stackoverflow{border-color:#a95009}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-steam{border-color:#006388}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-user_email{border-color:#b84e05}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-stumbleUpon{border-color:#9b280e}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-tumblr{border-color:#10151b}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-twitter{border-color:#0967a0}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-vimeo{border-color:#0d7091}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-windows{border-color:#003f71}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-whatsapp{border-color:#003f71}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-wordpress{border-color:#0f3647}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-yahoo{border-color:#14002d}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-youtube{border-color:#900}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-xing{border-color:#000202}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-mixcloud{border-color:#2475a0}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-vk{border-color:#243549}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-medium{border-color:#00452c}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-quora{border-color:#420e00}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-meetup{border-color:#9b181c}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-goodreads{border-color:#000}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-snapchat{border-color:#999700}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-500px{border-color:#00557f}.saboxplugin-socials.sabox-colored .saboxplugin-icon-color .sab-mastodont{border-color:#185886}.sabox-plus-item{margin-bottom:20px}@media screen and (max-width:480px){.saboxplugin-wrap{text-align:center}.saboxplugin-wrap .saboxplugin-gravatar{float:none;padding:20px 0;text-align:center;margin:0 auto;display:block}.saboxplugin-wrap .saboxplugin-gravatar img{float:none;display:inline-block;display:-moz-inline-stack;vertical-align:middle;zoom:1}.saboxplugin-wrap .saboxplugin-desc{margin:0 10px 20px;text-align:center}.saboxplugin-wrap .saboxplugin-authorname{text-align:center;margin:10px 0 20px}}body .saboxplugin-authorname a,body .saboxplugin-authorname a:hover{box-shadow:none;-webkit-box-shadow:none}a.sab-profile-edit{font-size:16px!important;line-height:1!important}.sab-edit-settings a,a.sab-profile-edit{color:#0073aa!important;box-shadow:none!important;-webkit-box-shadow:none!important}.sab-edit-settings{margin-right:15px;position:absolute;right:0;z-index:2;bottom:10px;line-height:20px}.sab-edit-settings i{margin-left:5px}.saboxplugin-socials{line-height:1!important}.rtl .saboxplugin-wrap .saboxplugin-gravatar{float:right}.rtl .saboxplugin-wrap .saboxplugin-authorname{display:flex;align-items:center}.rtl .saboxplugin-wrap .saboxplugin-authorname .sab-profile-edit{margin-right:10px}.rtl .sab-edit-settings{right:auto;left:0}img.sab-custom-avatar{max-width:75px;}.saboxplugin-wrap .saboxplugin-gravatar img {-webkit-border-radius:50%;-moz-border-radius:50%;-ms-border-radius:50%;-o-border-radius:50%;border-radius:50%;}.saboxplugin-wrap {margin-top:20px; margin-bottom:0px; padding: 0px 0px }.saboxplugin-wrap .saboxplugin-authorname {font-size:16px; line-height:23px;}.saboxplugin-wrap .saboxplugin-desc p, .saboxplugin-wrap .saboxplugin-desc {font-size:13px !important; line-height:20px !important;}.saboxplugin-wrap .saboxplugin-web {font-size:14px;}.saboxplugin-wrap .saboxplugin-socials a svg {width:18px;height:18px;}</style><script type='text/javascript' src='https://www.dataquest.io/wp-content/plugins/thrive-headline-optimizer/frontend/js/triggers.min.js?ver=1.2.7'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-includes/js/imagesloaded.min.js?ver=3.2.0'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-includes/js/masonry.min.js?ver=3.3.2'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-includes/js/jquery/jquery.masonry.min.js?ver=3.1.2b'></script>
<script type='text/javascript'>
/* <![CDATA[ */
var tve_frontend_options = {"is_editor_page":"","page_events":[],"is_single":"1","ajaxurl":"https:\/\/www.dataquest.io\/wp-admin\/admin-ajax.php","social_fb_app_id":"","dash_url":"https:\/\/www.dataquest.io\/wp-content\/plugins\/thrive-visual-editor\/thrive-dashboard","translations":{"Copy":"Copy"},"post_id":"2024","ip":"162.158.191.158","current_user":[],"post_title":"Jupyter Notebook for Beginners: A Tutorial","post_type":"post","post_url":"https:\/\/www.dataquest.io\/blog\/jupyter-notebook-tutorial\/","is_lp":"","post_request_data":[]};
var tve_frontend_options = {"ajaxurl":"https:\/\/www.dataquest.io\/wp-admin\/admin-ajax.php","is_editor_page":"","page_events":[],"is_single":"1","social_fb_app_id":"","dash_url":"https:\/\/www.dataquest.io\/wp-content\/plugins\/thrive-visual-editor\/thrive-dashboard","translations":{"Copy":"Copy","empty_username":"ERROR: The username field is empty.","empty_password":"ERROR: The password field is empty.","empty_login":"ERROR: Enter a username or email address.","min_chars":"At least %s characters are needed"},"routes":{"posts":"https:\/\/www.dataquest.io\/wp-json\/tcb\/v1\/posts"},"post_request_data":[],"ip":"162.158.191.158","current_user":[],"post_id":"2024","post_title":"Jupyter Notebook for Beginners: A Tutorial","post_type":"post","post_url":"https:\/\/www.dataquest.io\/blog\/jupyter-notebook-tutorial\/","is_lp":""};
/* ]]> */
</script>
<script type='text/javascript' src='https://www.dataquest.io/wp-content/plugins/thrive-visual-editor/editor/js/dist/frontend.min.js?ver=2.5.3.3'></script>
<script type='text/javascript' src='//www.dataquest.io/wp-content/plugins/thrive-leads/js/frontend.min.js?ver=2.2.14.3'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-content/plugins/posts-data-table/assets/js/datatables/datatables.min.js?ver=1.10.18'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-content/plugins/posts-data-table/assets/js/posts-data-table.min.js?ver=1.3'></script>
<script type='text/javascript'>
/* <![CDATA[ */
var tve_dash_front = {"ajaxurl":"https:\/\/www.dataquest.io\/wp-admin\/admin-ajax.php","force_ajax_send":"1","is_crawler":""};
/* ]]> */
</script>
<script type='text/javascript' src='https://www.dataquest.io/wp-content/plugins/thrive-visual-editor/thrive-dashboard/js/dist/frontend.min.js?ver=2.2.15.3'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-content/plugins/q2w3-fixed-widget/js/q2w3-fixed-widget.min.js?ver=5.1.9'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-includes/js/jquery/ui/core.min.js?ver=1.11.4'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-includes/js/jquery/ui/widget.min.js?ver=1.11.4'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-includes/js/jquery/ui/mouse.min.js?ver=1.11.4'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-includes/js/jquery/ui/resizable.min.js?ver=1.11.4'></script>
<script type='text/javascript'>
/* <![CDATA[ */
var thrive_front_localize = {"comments_form":{"error_defaults":{"email":"Email address invalid","url":"Website address invalid","required":"Required field missing"}},"routes":{"posts":"https:\/\/www.dataquest.io\/wp-json\/tcb\/v1\/posts"},"queried_object":{"ID":2024,"post_author":"7"},"tar_post_url":"?tve=true&action=architect&from_theme=1","is_editor":"","ID":"23111","template_url":"?tve=true&action=architect&from_tar=2024","pagination_href":"https:\/\/www.dataquest.io\/blog\/jupyter-notebook-tutorial\/page\/[thrive_page_number]\/","is_singular":"1"};
/* ]]> */
</script>
<script type='text/javascript' src='https://www.dataquest.io/wp-content/themes/thrive-theme/inc/assets/dist/frontend.min.js?ver=1.2.3'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-content/plugins/ank-prism-for-wp/out/prism-js.min.js?ver=1550607458'></script>
<script type='text/javascript' src='https://www.dataquest.io/wp-includes/js/wp-embed.min.js?ver=5.4.1'></script>
<script type="text/javascript">/*<![CDATA[*/var THO_Front = THO_Front || {}; THO_Front.data = {"end_of_content_id":"tho-end-content","is_single":true,"log_url":"https:\/\/www.dataquest.io\/wp-json\/tho\/v1\/logs","active_triggers":{"viewport":"thrive_headline"},"log_engagements":[],"post_id":2024,"test_id":0,"const":{"_e_click":1,"_e_scroll":2,"_e_time":3,"_impression":1,"_engagement":2}}/*]]> */</script><script class="tcb-post-list-template" type="text/template" data-identifier="tve-u-16eb7cc1dec">[tcb_post_list_dynamic_style]@media (min-width: 300px){.tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-[tcb_the_id].post-wrapper{background-image: linear-gradient(rgba(0, 0, 0, 0.55), rgba(0, 0, 0, 0.55)), url("[tcb_featured_image_url size=full]") !important;}:not(#tve) .tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-[tcb_the_id].post-wrapper:hover{background-image: linear-gradient(rgba(0, 0, 0, 0.5), rgba(0, 0, 0, 0.5)), url("[tcb_featured_image_url size=full]") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-[tcb_the_id] [data-css="tve-u-16eb7cc1def"]{background-image: url("[tcb_featured_image_url size=full]") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-[tcb_the_id] [data-css="tve-u-16eb7cc1dee"]:hover [data-css="tve-u-16eb7cc1def"]{background-image: linear-gradient(rgba(0, 0, 0, 0.05), rgba(0, 0, 0, 0.05)), url("[tcb_featured_image_url size=post-thumbnail]") !important;}}[/tcb_post_list_dynamic_style]
<a href="[tcb_post_the_permalink]" class="tve-dynamic-link" dynamic-postlink="tcb_post_the_permalink" data-shortcode-id="1929"><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-16eb7cc1dee" style="">
<div class="tve-content-box-background" data-css="tve-u-16eb7cc1def" data-tcb_hover_state_parent="1" style=""></div>
<div class="tve-cb" data-css="tve-u-16eb7cc1df0" style=""></div>
</div></a>
<div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-16eb7cd70e5"><p data-css="tve-u-16eb7cdd540"><strong>​</strong>​​​​<span class="thrive-shortcode-content" data-attr-date-format="F j, Y" data-attr-date-format-select="F j, Y" data-attr-show-time="0" data-attr-time-format="" data-attr-time-format-select="g:i a" data-attr-type="published" data-css="tve-u-17164265f0e" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_published_date" data-shortcode-name="Post Date" data-attr-link="0" data-attr-target="1" data-attr-rel="0" style="font-family: &quot;Open Sans&quot;; font-weight: 300;">[tcb_post_published_date date-format='F j, Y' date-format-select='F j, Y' show-time='0' time-format='' time-format-select='g:i a' type='published' inline='1' link='0' target='1' rel='0']</span></p></div>[tcb_post_title css='tve-u-1716426374a' tcb_hover_state_parent='']
</script><script class="tcb-post-list-template" type="text/template" data-identifier="tve-u-17146113c1f">[tcb_post_list_dynamic_style]@media (min-width: 300px){.tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-[tcb_the_id].post-wrapper{background-image: linear-gradient(rgba(0, 0, 0, 0.55), rgba(0, 0, 0, 0.55)), url("[tcb_featured_image_url size=full]") !important;}:not(#tve) .tve-theme-23111 .content-section [data-css="tve-u-17146113c1f"].tcb-post-list #post-[tcb_the_id].post-wrapper:hover{background-image: linear-gradient(rgba(0, 0, 0, 0.5), rgba(0, 0, 0, 0.5)), url("[tcb_featured_image_url size=full]") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-[tcb_the_id] [data-css="tve-u-16eb7cc1def"]{background-image: url("[tcb_featured_image_url size=full]") !important;}.thrive-section-23262 [data-css="tve-u-16eb7cc1dec"].tcb-post-list #post-[tcb_the_id] [data-css="tve-u-16eb7cc1dee"]:hover [data-css="tve-u-16eb7cc1def"]{background-image: linear-gradient(rgba(0, 0, 0, 0.05), rgba(0, 0, 0, 0.05)), url("[tcb_featured_image_url size=post-thumbnail]") !important;}}[/tcb_post_list_dynamic_style]
		<div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-17146113c23" style="">
	<div class="tve-content-box-background" data-css="tve-u-17146113c24" data-tcb_hover_state_parent="1"></div>
	<div class="tve-cb" data-css="tve-u-17146113c26"><div class="thrv_wrapper thrv_contentbox_shortcode thrv-content-box tve-elem-default-pad" data-css="tve-u-17146113c27">
	<div class="tve-content-box-background" data-css="tve-u-17146113c28" data-tcb_hover_state_parent="1"></div>
	<div class="tve-cb" data-css="tve-u-17146113c29"><div class="thrv_wrapper thrv_text_element" style="" data-css="tve-u-17146113c2a"><h2 class="" style="" data-css="tve-u-17146113c2b"><span class="thrive-shortcode-content" data-attr-link="1" data-attr-rel="0" data-attr-target="0" data-extra_key="" data-option-inline="1" data-shortcode="tcb_post_title" data-shortcode-name="Post title" data-css="tve-u-17146113c2c" style="" data-attr-css="">[tcb_post_title link='1' rel='0' target='0' inline='1' css='']</span></h2></div><div class="tcb-post-read-more thrv_wrapper" data-css="tve-u-17146113c2d" style="">
				<a href="[tcb_post_the_permalink]" class="tcb-button-link tcb-post-read-more-link">
					<span class="tcb-button-texts">
						<span class="tcb-button-text thrv-inline-text">​Read More</span>
					</span>
				</a>
			</div></div>
</div></div>
</div>
		

		

		

		
	</script><script type="text/javascript">var tcb_post_lists=JSON.parse('[{\"identifier\":\"[data-css=\\\"tve-u-16eb7cc1dec\\\"]\",\"template\":\"tve-u-16eb7cc1dec\",\"attr\":{\"query\":\"{&#039;paged&#039;:1,&#039;filter&#039;:&#039;custom&#039;,&#039;post_type&#039;:&#039;post&#039;,&#039;orderby&#039;:&#039;title&#039;,&#039;order&#039;:&#039;ASC&#039;,&#039;posts_per_page&#039;:&#039;4&#039;,&#039;offset&#039;:&#039;1&#039;,&#039;no_posts_text&#039;:&#039;There are no posts to display.&#039;,&#039;queried_object&#039;:{&#039;taxonomy&#039;:&#039;category&#039;,&#039;term_id&#039;:14},&#039;sticky&#039;:[],&#039;rules&#039;:[]}\",\"type\":\"list\",\"columns-d\":\"1\",\"columns-t\":\"2\",\"columns-m\":\"1\",\"vertical-space-d\":\"35\",\"horizontal-space-d\":\"35\",\"ct\":\"post_list-43427\",\"ct-name\":\"Image &amp; Text 01\",\"tcb-elem-type\":\"post_list\",\"pagination-type\":\"none\",\"pages_near_current\":\"2\",\"css\":\"tve-u-16eb7cc1dec\",\"article-tcb_hover_state_parent\":\"\",\"element-name\":\"Post List\",\"total_post_count\":\"299\",\"total_sticky_count\":0,\"posts_per_page\":4,\"featured-content\":\"0\",\"disabled-links\":\"1\",\"no_posts_text\":\"\",\"article-permalink\":\"[tcb-article-permalink]\",\"class\":\"\"},\"query\":{\"paged\":1,\"filter\":\"custom\",\"post_type\":\"post\",\"orderby\":\"title\",\"order\":\"ASC\",\"posts_per_page\":4,\"offset\":\"1\",\"no_posts_text\":\"There are no posts to display.\",\"queried_object\":{\"taxonomy\":\"category\",\"term_id\":14},\"sticky\":[],\"rules\":[]},\"posts\":[1929,2008,8505,2062]},{\"identifier\":\"[data-css=\\\"tve-u-17146113c1f\\\"]\",\"template\":\"tve-u-17146113c1f\",\"attr\":{\"query\":\"{&#039;paged&#039;:1,&#039;filter&#039;:&#039;custom&#039;,&#039;related&#039;:[],&#039;post_type&#039;:&#039;post&#039;,&#039;orderby&#039;:&#039;date&#039;,&#039;order&#039;:&#039;DESC&#039;,&#039;posts_per_page&#039;:&#039;4&#039;,&#039;offset&#039;:&#039;1&#039;,&#039;no_posts_text&#039;:&#039;There are no posts to display.&#039;,&#039;exclude_current_post&#039;:[&#039;1&#039;],&#039;queried_object&#039;:{&#039;ID&#039;:2437,&#039;post_author&#039;:&#039;1002&#039;,&#039;post_type&#039;:&#039;post&#039;},&#039;sticky&#039;:[],&#039;rules&#039;:[]}\",\"type\":\"grid\",\"columns-d\":\"2\",\"columns-t\":\"2\",\"columns-m\":\"1\",\"vertical-space-d\":\"19\",\"horizontal-space-d\":\"18\",\"ct\":\"post_list-43428\",\"ct-name\":\"Image &amp; Text 02\",\"tcb-elem-type\":\"post_list\",\"pagination-type\":\"none\",\"pages_near_current\":\"2\",\"css\":\"tve-u-17146113c1f\",\"article-tcb_hover_state_parent\":\"\",\"element-name\":\"Post List\",\"total_post_count\":\"298\",\"total_sticky_count\":0,\"posts_per_page\":4,\"featured-content\":\"0\",\"disabled-links\":\"1\",\"no_posts_text\":\"\",\"article-permalink\":\"[tcb-article-permalink]\",\"horizontal-space-m\":\"0\",\"vertical-space-m\":\"20\",\"horizontal-space-t\":\"15\",\"vertical-space-t\":\"15\",\"class\":\"\",\"article-class\":\"tve_evt_manager_listen tve_ea_thrive_animation tve_et_mouseover tve_anim_grow\"},\"query\":{\"paged\":1,\"filter\":\"custom\",\"related\":[],\"post_type\":\"post\",\"orderby\":\"date\",\"order\":\"DESC\",\"posts_per_page\":4,\"offset\":\"1\",\"no_posts_text\":\"There are no posts to display.\",\"exclude_current_post\":[\"1\"],\"queried_object\":{\"ID\":2437,\"post_author\":\"1002\",\"post_type\":\"post\"},\"sticky\":[],\"rules\":[]},\"posts\":[14526,23820,1943,23497]}]');</script><script type="text/javascript">/*<![CDATA[*/if ( !window.TL_Const ) {var TL_Const={"security":"6a8db6cedf","ajax_url":"https:\/\/www.dataquest.io\/wp-admin\/admin-ajax.php","forms":[],"action_conversion":"tve_leads_ajax_conversion","action_impression":"tve_leads_ajax_impression","ajax_load":1,"main_group_id":17187,"display_options":{"allowed_post_types":[],"flag_url_match":false},"shortcode_ids":["16321","19063"],"custom_post_data":{"http_referrer":"https:\/\/www.google.com\/"},"current_screen":{"screen_type":4,"screen_id":2024},"ignored_fields":["email","_captcha_size","_captcha_theme","_captcha_type","_submit_option","_use_captcha","g-recaptcha-response","__tcb_lg_fc","__tcb_lg_msg","_state","_form_type","_error_message_option","_back_url","_submit_option","url","_asset_group","_asset_option","mailchimp_optin"]};} else {ThriveGlobal.$j.extend(true, TL_Const, {"security":"6a8db6cedf","ajax_url":"https:\/\/www.dataquest.io\/wp-admin\/admin-ajax.php","forms":[],"action_conversion":"tve_leads_ajax_conversion","action_impression":"tve_leads_ajax_impression","ajax_load":1,"main_group_id":17187,"display_options":{"allowed_post_types":[],"flag_url_match":false},"shortcode_ids":["16321","19063"],"custom_post_data":{"http_referrer":"https:\/\/www.google.com\/"},"current_screen":{"screen_type":4,"screen_id":2024},"ignored_fields":["email","_captcha_size","_captcha_theme","_captcha_type","_submit_option","_use_captcha","g-recaptcha-response","__tcb_lg_fc","__tcb_lg_msg","_state","_form_type","_error_message_option","_back_url","_submit_option","url","_asset_group","_asset_option","mailchimp_optin"]})} /*]]> */</script>
</body>
</html>
````
